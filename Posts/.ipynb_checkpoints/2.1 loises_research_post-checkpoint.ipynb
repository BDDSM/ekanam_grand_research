## О том как мы будем достигать первую цель:

Как вы помните, первой целью иканам гранд рисёрча является поиск популярных, авторитетов, симпатий и интересного контента на основе анализа лайков и социальных связей. В текущей части исследования мы будем изучать лайки всех пользователей, которые принадлежат к эконому. Как выяснилось из прошлой части, таких пользователей около $700$. Лайки всех левых людей мы удалим. Статистика по лайкам собрана за два последних года. 

Как я уже говорил ранее, было бы интересно понять, кто из людей является хайпожором, кто постит самый годный контент, фотки и т.п. Лайки явно некоторым образом коррелируют с этими показателями. Наблюдая за лайками, мы попробуем достучаться до правды. Попутно мы построим довольно много графиков с разными лайк-топами. Более того,  для того, чтобы посмотреть на потоки лайков, мы построим в R интерактивную визуализацию, в которой можно будет узнать кого лайкает конкретный юзер и кто лайкает его. Ясное дело, что люди ведут себя в социальных сетках по-разному. Многие не ставят лайки вообще. Не забывайте о том, что лайк - всего лишь инструментальная переменная. Особенно когда будете играться с визуализацией.

## 3. Лайки, лайкопосты и лайкоматрицы

Построим **матрицу лайков!**  По строкам и столбцам расположим id людей. На месте $(i,j)$ будет стоять число лайков, которое пользователь $i$ поставил пользователю $j$. На диагонали окажутся самолайки. Обратите внимание на то, что в матрице никакой симметрии не будет. Построим такую матрицу по стенам пользователей, по фоткам и по всему вместе.

В дальнейшем нам хотелось бы понять какие люди популярны, то есть на регулярной основе делают лайкаемые в большом количестве посты. На нашем пути уже возникает преграда. Бывают пользователи, которые берут не качеством постов, а их числом. Пусть какая-то девочка на ежедневной основе заливает себе на стенку тпшные цитаты, а три её подруги ставят ей лайки. В конечном счёте за год она набирает больше $1000$ лайков. Пусть другая девочка постит в течение лета несколько фоток в купальнике. Каждая фотка срывает куш из лайков. В целом, фотки также набирают около $1000$ лайков. Если смотреть на абсолютные величины, обе девушки одинаково популярны, однако девушка в купальнике явно популярнее.

Одним словом, неплохо было бы отнормировать число лайков на количество постов и посмотреть ещё и на такой расклад. Будем называть отнормированные величины **лайкопостами.** В матрице лайкпостов будет стоять число лайков, поставленное пользователем $i$ пользователю $j$, поделённое на число постов пользователя $j$. Все числа будут расположены в диапазоне от 0 до 1. Чем ближе число к 1, тем **полнее** пользователь $i$ пролайкал объекты, которые выложил пользователь $j$.

Ещё раз, ещё раз. У нас есть две метрики.
* Лайки. Эта метрика будет выводить в топ тех, у кого много постов. Грубо говоря, штрафует за маленькое число постов и ищет хайпожоров.
* Лайкпосты. Эта метрика будет выводить в топ тех, у кого мало постов и все хорошо пролайканы. Грубо говоря, штрафует за большое число постов и ищет самых интенсивно залайканных малопостящих ребят. Обратите внимание на т

Нам бы хотелось найти золотую середину. Человек не постит очень много, при этом он хорошо залайкан. Интуиция подсказывает, что контент такого человека, скорее всего, является годным. Для того, чтобы найти таких людей, нам придётся усреднить лайки и лайкпосты и на их основе сделать новую метрику. Но об этом позже.

Матрица лайкопостов получается разряженой. Повсюду сплошные нули. Лайк — это редкое событие.

<img align="center" src="https://raw.githubusercontent.com/FUlyankin/ekenam_grand_research/master/images/2.%20loises_and_graph/matrix.png" height="500">

Можно построить аналогичную матрицу для более маленького чилса людей. Например, для рандомно отобранной двадцатки.

<img align="center" src="https://raw.githubusercontent.com/FUlyankin/ekenam_grand_research/master/images/2.%20loises_and_graph/20_matrix.png" height="500">

Матрица всё ещё разряжена. Такие же матрицы можно строить для конкретных людей, но это развлечение остаётся для тех, кто собирается вникать в код. Мы же займёмся визуализацией топов. Договоримся, что все картинки с топами, которые мы будем строить, будут содержать 30 человек.

## 5. Популярность, хайпожоры и интересный контент 

Итак, выше мы решили, что будем смотреть на лайки и лайкопосты. Посмотрим кто получил наибольшее число лайков в абсолютном выражении.

<img align="center" src="https://raw.githubusercontent.com/FUlyankin/ekenam_grand_research/master/images/2.%20loises_and_graph/all_top_likes.png" height="500">

Когда я впервые увидел этот график (для тех, кто не в курсе, Ульянкин Филипп, это я), я представил себя на церемонии вроде оскара, но только с золотыми лайками. Однако следующая картинка быстро сбила с меня спесь.

<img align="center" src="https://raw.githubusercontent.com/FUlyankin/ekenam_grand_research/master/images/2.%20loises_and_graph/all_top_likeposts.png" height="500">

Логично, что если человек вошёл в топ по лайкам, но не смог войти в топ по лайкпостам, в какой-то степени он берёт за счёт числа. Он хайпожор. Обратно, если человек вошёл в топ по лайкпостам, но не вошёл в топ по лайкам, скорее всего, он запостил несколько аватарок, которые собрали много лайков и ... всё.

В принципе, абсолютное количество лайков показывает нам какие люди являются популярными. Собрать большое число лайков даже при очень большом количестве постов - сложное занятие. Намного интереснее попытаться найти людей, которые постят не очень часто, но при этом довольно интересный контент, который собирает стабильно много лайков. Ни лайкпосты, ни лайки в силу своих недостатков, не могут вывести нас на такого человека. Такой человек лежит где-то в золотой середине. Его может найти только некоторое усреднение этих двух метрик, улавливающих противоположные вещи.

Поделим в матрице лайков все лайки на максимально возможное значение, чтобы все числа оказались в диапазоне от $0$ до $1$ и обе метрики были сопоставимы. Если мы возьмём от них среднее арифметическое

$$ A = \frac{1}{2} \cdot (\text{лайкпосты} + \text{лайки}), $$

то кривые безразличия между пользователями в пространстве двух метрик будут выглядеть как-то так:

<img align="center" src="https://raw.githubusercontent.com/FUlyankin/ekenam_grand_research/master/images/2.%20loises_and_graph/fmera1.png" width="400">

Прямыми на картинке обозначены кривые безразличия. По оси $x$ расположены лайки (precision), по оси $y$ лайкпосты (recall). Такие странные названия будут пояснены позже. Представим себе двух юзеров. Первый очень мало постит и в абсолютном выражении получает довольно мало лайков, но при этом получает много лайкпостов. Второй юзер постит среднее количество постов, которое покрыто средним числом лайков. Логично, что второй человек должен оказаться в итоговом рейтинге выше первого. В нашей же ситуации вполне может возникнуть ситуация, когда оба юзера окажутся на одной и той же кривой безразличия. Именно такая ситуация изображена на рисунке. 

Исправить это можно несколькими разными путями. Например, можно попробовать брать минимум из двух метрик. Как будут в этом случае выглядеть кривые безразличия, вы легко можете угадать сами...

<img align="center" src="https://raw.githubusercontent.com/FUlyankin/ekenam_grand_research/master/images/2.%20loises_and_graph/fmera15.png" width="350">

Однако это также не самый хороший вариант. Два человека с одинаковыми лайками, но разными лайкпостами оказываются эквивалентными. Тем не менее, человек с большим числом лайкпостов явно круче. Кривые безразличия хорошо бы сделать выпуклыми. Именно это позволяет сделать штука, которая носит гордое имя **f-мера** и задаётся следующим образом:

$$ F = (1 + \beta^2) \cdot \frac{\text{лайки} + \text{лайкпосты}}{\beta^2 \cdot \text{лайки} + \text{лайкпосты}} $$

Линии уровня, которые соотвествуют f-мере, будут выпуклыми.

<img align="center" src="https://raw.githubusercontent.com/FUlyankin/ekenam_grand_research/master/images/2.%20loises_and_graph/fmera2.png" width="400">

Обратите внимание, что хайпожоры будут находиться в правом нижнем углу, а люди с малым количеством постов в левом верхнем. Оба угла находятся на низких кривых безразличия.

Выбирая параметр $\beta$, можно отдавать предпочтения либо лайкам либо лайкпостам. Например, на картинке выше, важнее лайкпосты. Если поставить большое значение $\beta$, например  $\beta = 2$, важнее будут лайки. 

Вы парень и ваша девушка истеричка, которая написала код на питоне, мониторящий ваши действия в соц сети? Откалибруйте её $\beta$ так, чтобы на первом месте в вашем топе всегда была она. Не забудьте после этого поблагодарить математику за решение вашей проблемы.

На самом деле все проделанные выше построения в базовых курсах машинного обучения делаются над метриками recall (полнота) и precision (точность). Полученную на их основе f-меру довольно часто используют при решении задачи классификации. Однако об этом мы поговорим в самом конце нашего рисёрча. Запомните эти примеры с лайками и в будущем сопоставьте с ними всю ту математику, на которую мы посмотрим.

## 14. Немного итогов

Подведём промежуточные итоги. Мы хотели: 

* Найти популярных людей
* Найти людей с годным контентом 
* Найти взаимные симпатии 
* Найти самых авторитетных людей 

В этой части мы попробовали найти популярных людей по количеству полученых лайков. Это не принесло нам хороших результатов, так как люди могут завоёвывать популярность за счёт числа постов. Чтобы решить эту проблему, мы построили лайкпосты, нечувствительные к числу оставленых постов. К сожалению, это привело к новым проблемам. В топ стали выбиваться люди с маленьким числом постов, но большим числом лайков под ними. Нам же хотелось бы найти человека, который постит более-менее стабильно. В конечном счёте нашей первой метрикой, измеряющей популярность, стала f-мера, которая может учесть в себе оба эффекта: и полноту пролайканости стенки, и число лайков, оставленное на ней. 

Альтернативным f-мере, более простым показателем, для поиска популярных людей с хорошим контентом, стали продискантированные лайки. Мы решили, что лайки людей, которые лайкают редко, являются более ценными. В связи с этим, мы взвесили число лайков, которое оставляет человек, на общее число лайков, которое он оставил в течение двух лет. Это позволило учитывать лайки от редколайкающих людей с более большим весом. 

Поиск людей, которые чаще всего попадают к другим ребятам в топ-1 по числу оставленных лайков, стал третьей метрикой, отвечающей за популярность. Результаты топов в какой-то степени близки. Какому из них больше верить, выбирать вам. 

В самом конце исследования мы попытались идентифицировать пары людей, которые находятся друг у другу в топе по числу лайков. Эти мэтчинги вышли довольно забавными. Особенно мэтчинги для топ-2, которые вы, конечно же, уже посмотрели в [файлике с кодом.](http://nbviewer.jupyter.org/github/FUlyankin/ekanam_grand_research/blob/master/2.Graph_and_loises/2.1%20loises_research%20.ipynb) Про авторитетность в этом файле мы не сказали ни слова. Её исследование нас ожидает в следующей статейке.
