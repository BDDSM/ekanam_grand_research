# 1. Про ошибки, которые мы допускаем 

Скачать контент всех пабликов, а после скачать все лайки, которые стоят под каждым постом, дело нетрудное, но времязатратное. Около 6 часов уйдёт на то, чтобы скачать из 20000 пабликов по 100 постов. После надо будет к каждому из 200000 постов обратиться по несколько раз (в зависимости от числа лайков), чтобы собрать списки из id, оставивших своё мнение под постом. На это, при использовании метода `execute`, который, грубо говоря, позволяет делать запросы в 25 раз быстрее, уйдёт около 18 часов.

Код для парсинга получится совсем небольшим. Когда он будет дописан, вы включите свой старый ноут, поставите на нём код (новый ноут вы возьмёте с собой на пары/работу/тусу), увидите вот такую, радующую глаз картину и оставите расчёты на произвол судьбы. 

<img align="center" src="../images/3. publics/likes.png" width="1200">



Вроде бы код работает нормально, все проверки показали, что он, действительно, скачивает разные лайки с разных групп, под разными постами. Вы однажды уже ставили расчёты на 6 часов, а потом неожиданно обнаружили, что код скачал $20000$ раз одно и то же. Именно тогда вы прокричали своё **первое даташрушерское блять.** Сегодня вы такого не допустите. Код проверен много раз и работает корректно. Интересно, что же мы увидим через $18$ часов...

Мы увидим зависший экран. Неожиданно обнаружится, что вашему старому ноуту уже 6 лет, на нём очень мало оперативки и с новым маком, на котором вы готовили этот код, он соперничать в плане производительности не может. Всё зависло на $20\%$. НУ КОНЕЧНО! Надо было внутри кода прописать, чтобы данные сохранялись в новый файлик каждые $10\%$. Тогда бы это зависание не было бы таким обидным, но вы это не сделали. Вы рассчитывали, что эти $18$ часов будут отработаны кодом также чётко, как были отработаны предыдущие $6$. Вместо этого всего вы уже орёте во всё горло: **БЛЯТЬ!** 

Не беда, завтра же выходной. Вы весь день планировали протусить дома. Можно запустить расчёты на маке на всю ночь и весь день. Перезапускаем код на маке, убеждаемся что всё окей и, когда счётчик доходит до $5\%$, отправляемся спать. Вот смеху то будет утром, когда вы проснётесь и неожиданно выясните, что забыли обновить ключ доступа к API. А он, между прочим, выписывается на $24$ часа. Особенно забавно то, что он истёк ровно через $10$ минут после того как вы уснули. Весь экран заполонит полчище сообщений в стиле `Проблемы с 228 батчем`. Что будет дальше? Вы, как настоящий дата сайнтист, проорёте во всё горло то самое сокровенное и неистовое **БЛЯТЬ,** обновите ключ доступа и перезапустите код. 

Третья попытка увенчается успехом. Всё скачалось. Гигабайты из лайков лежат на жёстком диске вашего компьютера. Вместо одних суток, которые вы планировали потратить на скачку, вы потратили трое. Это только скачка. Данные надо будет ещё обработать, после надо будет обучить несколько моделей и получить какие-то результаты. Конечно же вы попытаетесь предотвратить все эксцессы и сначала попробуете запускать код на маленьких подвыборках, пока не окажетесь уверены в своём коде на все $99%$. # 3. Подготовка данных к кластеризации

Перед тем как строить по корпусу из текстов каки-либо модели, нужно превратить тексты в цифры. Математика работает только с цифрами, а моделирование тем пабликов ещё та математика. Во-первых, сделаем все буквы строчными. Во-вторых, разрубим все тексты на состовляющие их слова. Такую разрубку называют **токенизацией.** 

Обычно, при работе с текстами руководствуются гипотезой, которая называется **мешком слов.** Смысл этой гипотезы состоит в том, что порядок слов в предложении никак не сказывается на его смысле. 

* Порядок слов неважен
* Неважен порядок слов
* Слов порядок неважен 
* Порядок неважен слов 
* Слов неважен порядок
* Неважен слов порядок 

Конечно же с этой гипотезой можно поспорить и для этого вам придётся схлеснуться в словесной дуэли с одним известным персонажем. 

<img align="center" src="http://tvperson.ru/2014/11/yoda-3-1024x682.jpg" width="500"> 

Главным аргументом с вашей стороны, скорее всего, будет частица не, положение которой в предложении может координально влиять на смысл фразы. Да, порядок слов иногда важен. Тем не менее при моделировании мы должны чем-то жертвовать. Справедливости ради стоит отметить, что существует ряд более сложных моделей, которые учитывают порядок слов, но мы не будем их касаться в этом исследовании. По крайней мере пока. Тем более, интуиция подсказывает, что пренебрежение порядком слов не должно никак сказаться на выделении тем.

Грубо говоря, каждое слово — это регрессор. Перед каждым словом надо оценивать коэффициент. То с какой частотой оно наблюдается в статье — наше наблюдение. Наблюдений в сравнении с числом параметров, которое нужно будет оценить, очень мало. Постараемся уменьшить число регрессоров, выкинув не очень важные слова.

Например, в текстах часто встречаются предлоги. Они не несут никакой смысловой информации о сути текста и есть почти в каждой группе. Найдётся огромное количество подобной лексики, которая для нашего набора текстов встречается повсеместно и не несёт никакой информации. Обычно такие частые слова называют **стоп-словами.** Они путают нашу модель, перетягивая на себя внимание. От них необходимо избавиться.

Другим естественным путём для уменьшения числа регрессоров является приведение всех слов к одной форме. Всех существительных к именительному падежу и единственному числу, всех глаголов к инфинитиву и т.д. Такая обработка называется **лемматизацией.** Она делается по специальным словарям, составленным для языка. К счастью, для русского языка такой словарь реализован в ультракрутом пакете ['pymorthy2'.](http://pymorphy2.readthedocs.io/en/latest/)

Этот пакет превратит слово люди в слово человек, слова красивая и красивые в слово красивый, слова сделал и делал в делать. Таким образом количество различных слов в словаре здорово упадёт, а смысл не пострадает.

# 4. Что такое тематическое моделирование

Пришло время обсудить то, как именно рождаются тексты. Разговаривать об этом мы будем не особо глубоко. Нашей задачей является понять то, что именно нам выдаст на выходе алгоритм. То какие именно тонкости за ним стоят и как происходит оценивание, мы опустим. Эти тонкости очень важны, но вы читаете не скучный учебник по машинному обучению, а увлекательный гранд иканам ресёрч. Поэтому мы позволим себе такую вольность и не будем говорить ни о методе максимального правдоподобия, ни о четырёх новых методах наименьших квадратов и всё такое. Обо всём этом вы поговорите на курсе или факультативе по анализу данных. Здесь же мы будем говорить о мешках и совсем немного о матрицах. Тем не менее если вы учитесь на первых курсах, иногда, число незнакомых слов будет зашкаливать. Крепитесь.

Итак, отбросим прочь все предрассудки и глупости. Наши мир создал он, великий процесс порождения данных! И каждый текст он создал в виде мешка. Мешка слов. На самом деле мешков много. И каждый мешок соответствует своей теме. Внутри мешков лежат камушки. Много камушков. На каждом камушке написано какое-то слово. Слова часто повторяются. Каждое слово повторяется разное количество раз. 

<img align="center" src="../images/3. publics/meshok_1.png" width="600"> 


Дальше природа подкидывает кубик с количеством граней равным количеству тем и выбирает одну из них. Кубик у природы неправильный. Разные грани выпадают на нём с разными вероятностями. Из мешка, который соответствует этой теме, природа тянет камушек.

<img align="center" src="../images/3. publics/meshok_2.png" width="450"> 

После природа подкидывает кубик снова, выбирает какую-то новую тему и начинает тянуть из мешка, который ей соответствует, новый камушек. Так и рождается текст.

Поговорим о порождении природой текстов чуть более научно. Будем называть каждый текст **документом, $d$.** Каждый документ состоит из нескольких **тем, $t$.** Каждая тема описывается некоторым набором **слов, $w$.**

Как мы выяснили выше, каждый документ — это мешок слов. Для каждого слова мы можем посчитать с какой частотой оно входит в каждый документ. Эта частота будет оценкой вероятность встретить в документе конкретное слово, $p(w | d)$. В конечном счёте, мы видим матрицу, где по строкам откладываются конкретные слова, по столбцам документы, а на пересечении стоят частоты, с которыми наши слова встречаются в документах.

<img align="center" src="../images/3. publics/matrix1.png" width="300"> 


Мы хотели бы научиться выделять в каждом документе темы. Руководствуясь этим желанием, мы можем сказать, что процесс порождения текстов описывается следующим образом. Есть какое-то, изначально задуманной природой, количество тем, $T$. Для каждого документа существует некоторое распределение тем в нём, $p(t | d) = \theta_{td}$. Природа случайно выбирает из него одну из тем. После из распределения слов $p(w | t) = \phi_{wt}$, которое соответствует выбранной теме, выбирается слово. Именно оно и попадает в текст.

<img align="center" src="../images/3. publics/doc_generate.png" width="1100"> 

Обратите внимание, что каждое слово втыкается в документ независимо от того какие слова были туда запиханы до него (мешок слов). Греческие буквы для обозначения этих вероятностей используются для удобства.

В конечном счёте получается, что мы можем встретить слово $w$ в документе $d$, если оно выскочило к нам из темы $t_1$ или из темы $t_2$ и так далее. Возникает $T$ гипотез возникновения слова в документе. Полная вероятность появления слова может быть вычислена по формуле полной вероятности.

$$
p(w | d) = \sum_{t \in T} p(w | t) \cdot p(t |d) = \sum_{t \in T} \phi_{wt} \cdot \theta_{td}
$$

Вся матрица слова на документы таким образом может быть разложена в произведение из двух более маленьких матриц. 


<img align="center" src="../images/3. publics/matrix.png" width="700"> 

В матрице $\Phi$ по строкам расположены слова, по столбцам темы. На пересечении стоит вероятность встретить слово из строки в теме из столбца. В матрице $\Theta$ по строкам находятся темы, по столбцам документы. На пересечении стоит вероятность встретить тему из строки в документе из столбца. При перемножении каждой строки на каждый столбец, получается ровно требуемая нам полная вероятность встретить слово $w$ в документе $d$.

Если говорить обо всём этом в терминах мешков, в матрице $Θ$ лежат вероятности вытянуть камушки с конкретными словами. В матрице $Φ$ лежат вероятности, с которыми кубик выпадает разными гранями.

Вероятности из матриц $\Phi$ и $\Theta$ неизвестны. Наша задача сводится к тому, чтобы как-то их оценить. Такая задача носит название **задачи матричного разложения.** Человечество неплохо умеет её решать. Если говорить совсем грубо, то на неизвестные коэффициенты матриц накладываются ограничения, дальше выписывается целевая функция и лагранджиан для неё. Потом ~~находят первокурсника~~ минимизацию делают с помощью градиентного спуска.

Именно на решении такой задачи основаны многие рекомендательные системы в интернет-магазинах и на сайтах. Если говорить чуть менее грубо, то, обычно, всё сводится к максимизации правдоподобия. Более того, есть даже специальные методы наименьших квадратов, которые решают эту задачу.

Стоит сказать, что сама по себе, задача матричного разложения не имеет единственного решения. Если между нашими матрицами впихнуть ещё одну, мы получим новое решение.

$$ ΦΘ = (ΦS)(S^{−1}Θ) = Φ^1Θ^1 $$

Из-за этого приходится вводить какие-то дополнительные ограничения на структуру матриц. Введение таких ограничений называется **регуляризацией.** На самом деле есть разные ограничения на коэффициенты матриц. Одни из них пытаются сделать темы независимее друг от друга, другие пытаются выделить фоновую лексику, третьи пытаются сделать матрицы более разряженными. Для всего этого есть даже [ультра крутая библиотека ARTM.](https://www.youtube.com/watch?v=H7hlSz4WWhQ)
Самым знаменитой регуляризацией является регуляризация Дирихле, которую придумали в 2003 году. Именно её мы и будем использовать.

** Ещё раз, ещё раз:** Мы хотим оценить вероятность появления каждого слова в каждой теме и каждой темы в каждой группе. Число тем мы будем задавать сами. Для решения этой задачи мы будем использовать модель с регуляризатором Дирихле, которая носит имя LDA. Попробуем разбить корпус из групп на 20 разных кластеров.

````
['лого', 'коллаб', 'supreme', 'четверг', 'худи', 'бокс', 'расцветках', 'превью', 'футболки', 'вещей', 'расцветки', 'куртку', 'футболка', 'принтом', 'суприм', 'сезон', 'куртка', 'лонгслив', 'брэнда', 'брэндинг']
```

Десятая тема оказывается телевизионно-киношной.

```
['коллабе', 'фильм', 'hanes', 'матч', 'сезона', 'жанр', 'команды', 'фильма', 'сезон', 'news', 'кино', 'серия', 'фадеев', 'madrid', 'hala', 'мемасиков', 'перевеса', 'серии', 'закупайтесь', 'йоркски']
```

Двенадцатая тема отвечает за бизнес. 

```
['бизнес', 'цели', 'бизнеса', 'будете', 'цель', 'клиентов', 'правило', 'идея', 'способ', 'результат', 'идеи', 'успеха', 'ситуации', 'решения', 'уровень', 'заниматься', 'книга', 'читать', 'план', 'процесс']
```

Пятнадцатая тема отвечает за духовность и государство. 

```
['аллах', 'аллаха', 'россия', 'войны', 'власти', 'news', 'ссср', 'территории', 'республики', 'президента', 'стране', 'государства', 'путин', 'президент', 'населения', 'армии', 'российской', 'украины', 'заявил', 'глава']
```

Семнадцатая тема оказывается связана с чем-то ванильным 

```
['мама', 'знаешь', 'тобой', 'твои', 'девушка', 'говно', 'папа', 'родители', 'твой', 'парень', 'домой', 'напиши', 'можешь', 'хотелось', 'надеюсь', 'доме', 'спать', 'вроде', 'давай', 'понимаю']
```

В нашей группе доминирует телевизор. Судя по всему, она связана либо с кино, сериалами либо с футболом. Внимание, правильный ответ: нашей группой оказывается `Зенит | FC Zenit`.

Отлично! Попробуем сделать то же самое ещё с парой групп. Внимательно посмотрите на темы, вероятности, с которыми эти темы встречаются в группах и на слова, которые вероятнее всего характеризуют каждую тему и попробуйте догадаться что за группа перед вами. 

-------------

### Человек против LDA. Раунд первый. 

Распределение тем в группе номер $331$: 

```
(12, 0.02803703), (13, 0.17606306), (17, 0.78327233)
```

Слова, которые описывают соответствующие темы: 

```
тема 12: ['бизнес', 'цели', 'бизнеса', 'будете', 'цель', 'клиентов', 'правило', 'идея', 'способ', 'результат', 'идеи', 'успеха', 'ситуации', 'решения', 'уровень', 'заниматься', 'книга', 'читать', 'план', 'процесс'] 
```

```
тема 13: ['женщина', 'женщины', 'однажды', 'смерти', 'словно', 'женщин', 'тело', 'бога', 'мужчина', 'мужчины', 'смерть', 'сердца', 'чувства', 'отношения', 'земле', 'солнце', 'душа', 'отец', 'любить', 'мать'] 
```

```
тема 17:  ['мама', 'знаешь', 'тобой', 'твои', 'девушка', 'говно', 'папа', 'родители', 'твой', 'парень', 'домой', 'напиши', 'можешь', 'хотелось', 'надеюсь', 'доме', 'спать', 'вроде', 'давай', 'понимаю'] 
```

У вас есть пара минуток на размышление. 

**Правильный ответ:** Основное внимание нужно обратить на 13 и 17 кластеры. Слова из 12 кластера в контенте группы представлены в довольно маленьком количестве. Судя по всему, группа про отношения, семью и сопли. Этой группой оказываеcя группа с эпичным названием 'Моя душа.🌸' 


---------------------

### Человек против LDA. Раунд второй. 

Распределение тем в группе номер $1110$:

```
(6, 0.025086233), (12, 0.71532333), (13, 0.016799694), (17, 0.20618792
```

Выведем слова только для самых вероятных тем. 

```
тема 12: ['бизнес', 'цели', 'бизнеса', 'будете', 'цель', 'клиентов', 'правило', 'идея', 'способ', 'результат', 'идеи', 'успеха', 'ситуации', 'решения', 'уровень', 'заниматься', 'книга', 'читать', 'план', 'процесс'] 
```

```
тема 17:  ['мама', 'знаешь', 'тобой', 'твои', 'девушка', 'говно', 'папа', 'родители', 'твой', 'парень', 'домой', 'напиши', 'можешь', 'хотелось', 'надеюсь', 'доме', 'спать', 'вроде', 'давай', 'понимаю'] 
```

И снова пара минут на размышление. 

**Правильный ответ:** Нет, это не бизнес-молодость. Но чем-то похоже. Паблик называется 'Жизнь Без Диплома'.

---------------------

### Человек против LDA. Раунд третий. 

Распределение тем в группе номер $10101$:

```
(6, 0.83615762), (13, 0.088842362)
```

Выведем слова только для самых вероятных тем. 

```
тема 6: ['студентов', 'бизнес', 'курс', 'развития', 'программы', 'студенты', 'регистрация', 'обучения', 'университета', 'курса', 'обучение', 'компаний', 'проектов', 'управления', 'подробнее', 'конференции', 'образования', 'данных', 'сфере', 'российской'] 
```

И снова пара минут на размышление. 

**Правильный ответ:** Перед вами паблик, который называется 'Институт повышения квалификации'.


---------------------

### Человек против LDA. Раунд четвёртый. 

Распределение тем в группе номер $17008$:

```
(4, 0.98398751)
```

Для данного паблика всё оказалось слишком конкретным...

```
тема 4: ['ингредиенты', 'масло', 'добавить', 'соль', 'приготовление', 'тесто', 'яйца', 'вкусу', 'перец', 'масла', 'сахар', 'салат', 'ложки', 'приятного', 'нарезать', 'сахара', 'мясо', 'аппетита', 'перемешать', 'молоко'] 
```

**Правильный ответ:** 'Рецепты'.

---------------------

### Человек против LDA. Раунд пятый. 

Распределение тем в группе номер $2528$:

```
(10, 0.05097561), (12, 0.056773484), (17, 0.63371867), (19, 0.23094608)
```

Тут получается немного забаностей. 
```
тема 13: ['женщина', 'женщины', 'однажды', 'смерти', 'словно', 'женщин', 'тело', 'бога', 'мужчина', 'мужчины', 'смерть', 'сердца', 'чувства', 'отношения', 'земле', 'солнце', 'душа', 'отец', 'любить', 'мать'] 


```

**Правильный ответ:** 'Рецепты'.

---------------------

### Человек против LDA. Раунд шестой. 

Распределение тем в группе номер $1726$:

```
(0, 0.27144378), (1, 0.029365834), (6, 0.24676128), (13, 0.029583057), (15, 0.38688779)
```

Тут получается немного забаностей. 
```
тема 0: ['мастер', 'вход', 'состоится', 'конкурса', 'ждем', 'конкурс', 'билеты', 'сентября', 'москвы', 'поздравляем', 'приходите', 'мероприятия', 'концерт', 'фестиваля', 'фестиваль', 'санкт', 'пройдет', 'команда', 'праздник', 'приглашаем'] 

тема 6: ['студентов', 'бизнес', 'курс', 'развития', 'программы', 'студенты', 'регистрация', 'обучения', 'университета', 'курса', 'обучение', 'компаний', 'проектов', 'управления', 'подробнее', 'конференции', 'образования', 'данных', 'сфере', 'российской'] 

тема 15: ['аллах', 'аллаха', 'россия', 'войны', 'власти', 'news', 'ссср', 'территории', 'республики', 'президента', 'стране', 'государства', 'путин', 'президент', 'населения', 'армии', 'российской', 'украины', 'заявил', 'глава'] 

```

Что же будет если взять студактив, добавить к нему немного учёбы и приправит это всё государством/религией? 

**Правильный ответ:** Конечно же получится 'ЕДИНАЯ РОССИЯ'.

---------------------

### Человек против LDA. Раунд седьмой. 

Распределение тем в группе номер $1815$:

```
(0, 0.092384227), (1, 0.017113004), (2, 0.085320123), (3, 0.01131352), (6, 0.10741896),
(12, 0.12030204), (13, 0.081164517), (15, 0.47009802)
```

Тут получается немного забаностей. 
```

тема 12: ['бизнес', 'цели', 'бизнеса', 'будете', 'цель', 'клиентов', 'правило', 'идея', 'способ', 'результат', 'идеи', 'успеха', 'ситуации', 'решения', 'уровень', 'заниматься', 'книга', 'читать', 'план', 'процесс'] 

тема 15: ['аллах', 'аллаха', 'россия', 'войны', 'власти', 'news', 'ссср', 'территории', 'республики', 'президента', 'стране', 'государства', 'путин', 'президент', 'населения', 'армии', 'российской', 'украины', 'заявил', 'глава'] 

```

А если взять политику/религию и добавит в неё немного бизнеса? 

**Правильный ответ:** Конечно же получится 'Фонд борьбы с коррупцией (ФБК)'.

---------------------

### Человек против LDA. Раунд восьмой. 

Распределение тем в группе номер $299$:

```
(13, 0.24010138), (15, 0.74209112)
```

Тут получается немного забаностей. 
```

тема 13: ['женщина', 'женщины', 'однажды', 'смерти', 'словно', 'женщин', 'тело', 'бога', 'мужчина', 'мужчины', 'смерть', 'сердца', 'чувства', 'отношения', 'земле', 'солнце', 'душа', 'отец', 'любить', 'мать'] 

тема 15: ['аллах', 'аллаха', 'россия', 'войны', 'власти', 'news', 'ссср', 'территории', 'республики', 'президента', 'стране', 'государства', 'путин', 'президент', 'населения', 'армии', 'российской', 'украины', 'заявил', 'глава'] 

```

Теперь смешаем между собой ванильную тему и государство/религию. Что получится? 

**Правильный ответ:** Конечно же получится 'Ислам - это Коран и  Сунна'.

--------------------

Вроде бы темы получаются очень даже интерпретируемыми. Обратите внимание, что некоторые темы явно слиплись. Например, государство слиплось с религией.  Это сигнал к тому, что надо увеличить число кластеров и построить более большую модель. Но это немного позже. 

# 7. Оценка качества кластеризации 

Кластеризацию провели, на группы посмотрели. Дело осталось за малым. Понять насколько хорошо работает построенная нами модель.  Обычно, в случае кластеризации, это довльно сложно.  К сожалению, мы не знаем правильных ответов на вопрос сколько всего существует групповых тематик, а уж тем более, мы не знаем какая группа к какой тематике относится. 

* **Экспертная оценка**.

Экспертам предлагают оценить насколько выделенные темы интерпретируемы. Иногда видно, что в темах нет никакого смысла или что какие-то несколько тем слипаются в одну и количество кластеров неплохо было бы увеличить. 

Например, экспертам предлагается рассмотреть темы как последовательности слов, упорядоченные по вероятности встретить слово в данной теме, и оценить интерпретируемость темы по некоторой шкале (2 или 5-бальной). Обычно экспертам дают такую инструкцию: если эту тему можно кратко озаглавить или по этим словам можно сформулировать поисковый запрос и получить релевантную поисковую выдачу, то такую тему следует считать интерпретируемой. Такие оценки, естественно, являются субъективными, поэтому каждую тему оценивают несколько экспертов, чтобы понять, насколько непротиворечивы их оценки.

Другой метод, так называемый метод интрузий (intrusion), заключается в том, что в список топовых (имеющих большую вероятность) слов темы внедряется лишнее слово, которое заведомо этой теме не принадлежит. Экспертам предлагается определить, какое слово из списка лишнее, и измеряется доля ошибок при его определении. Такой способ оценки интерпретируемости существенно проще для экспертов, а, следовательно, эксперт может оценить больше тем в единицу времени.

* **Перплексия**. 

Эта мера качества тесно связана с правдоподобием. По сути дела это экспонента от усредненного по всем словам всех документов значения логарифма правдоподобия. Можно сказать, что это мера неопределенности или различности слов в тексте. Если распределение слов неравномерно, то перплексия уменьшается по сравнению с тем значением, которое дает равномерное распределение. Еще можно сказать, что перплексия — коэффициент ветвления текста, то есть количество ожидаемых в среднем различных слов после каждого слова в документе. Чтобы оценить перплексию, можно использовать отложенную выборку. Однако эксперименты на больших коллекциях показывают, что различие между перплексией на обучающей и на тестовой выборках, как правило, не существенно для цели сравнения разных моделей.

Перплексия ничего не говорит об интерпретируемости тем, а только говорит о том, насколько хорошо построилось матричное разложение. То есть ничего не говорит о том, насколько построенная модель будет полезна для конечных приложений. Более того, эта метрика зависит от размера словаря и непригодна для сравнения моделей с разными словарями. 

* **Когерентность**.

В ходе экспериментов было выявлено, что экспертные оценки хорошо коррелируют с такой мерой качества как когерентность, которая может быть вычислена полностью автоматически без участия человека. Когерентность темы — мера, которая показывает, насколько слова, встречающиеся рядом в текстах, оказываются в топах одних и тех же тем. 

Окей! Попробуем взглянуть на то, что у нас получается по нашей модели. В первую очередь будем ориентироваться не на метрики, а на свой здравый, экспертный смысл. 

Посмотрим ещё раз на самые вероятные слова, которые встречаются в темах и на группы, которые с самой высокой вероятностью к этой теме относятся. Чтобы было повеселее, можно даже придумать для каждой из тем своё название. Если такое название практически сразу приходит в голову, значит тема интепретируется и выделена адекватно. Если нет, то кластеризация оказалась не очень хорошей. Длинный список из вероятностей и тестов закончится котейкой. 


    ['мастер', 'вход', 'состоится', 'конкурса', 'ждем', 'конкурс', 'билеты', 'сентября', 'москвы', 'поздравляем', 'приходите', 'мероприятия', 'концерт', 'фестиваля', 'фестиваль', 'санкт', 'пройдет', 'команда', 'праздник', 'приглашаем']
    0.991518 : Relax Art ™
    0.983338 : S.O.V.A. - #ГЛАЗАОТКРОЙ
    0.980636 : jane's shots
    0.926771 : Капитаны
    0.925063 : Студенческое Самоуправление ИАТУ УлГТУ
    0.923066 : КУРИЛЫ (Профсоюзная)
    0.920659 : ГДОД "ЮНАЯ ТУЛА"
    0.916942 : Афиша Нижневартовска
    0.914861 : КУРИЛЫ (ВДНХ)
    0.914418 : СТУДЕНТ-ЭКОНОМИСТ | Смешные Мемы и Анекдоты
    0.904263 : TODES/Тодес
    0.904115 : Школа танцев ФАБРИКА ТАНЦЕВ
    0.901919 : Посвящение 2014 Владимирского филиала РАНХиГС
    0.899865 : Свободное Пространство «ТИПИЧНЫЙ ПИТЕР» Антикафе
    0.894444 : Фу, тамблер
    0.88125 : EGE-GO - ОГЭ КДР ВПР Ответы (Запасной паблик)
    0.88125 : ФГБУ «МНИИ ГБ им.Гельмгольца» Минздрава России
    0.880822 : elcentro
    0.877166 : Парк Культуры и Отдыха " Ёлочки"
    0.875347 : sweet home


    ['лого', 'коллаб', 'supreme', 'четверг', 'худи', 'бокс', 'расцветках', 'превью', 'футболки', 'вещей', 'расцветки', 'куртку', 'футболка', 'принтом', 'суприм', 'сезон', 'куртка', 'лонгслив', 'брэнда', 'брэндинг']
    0.741547 : Журнал "Супримилка"
    0.572737 : Модный AliExpress/Поиск Товара
    0.560206 : AliExpress Хайповый Шмот
    0.528045 : AliExpress Хайповый Шмот
    0.525 : minimalism
    0.525 : Twitter Dota
    0.508333 : ДГТУ_МТиО_ОМ-23_2014.
    0.473419 : Vk канал : " Дарья Мэг "
    0.454771 : Модели и фотографы
    0.453096 : NOMODELS
    0.428649 : Al 6 Medwedsky
    0.424521 : ЧТО БЫ ТЫ ВЫБРАЛ?
    0.422646 : apparelVIBES
    0.419171 : Институт управления,бизнеса и технологий
    0.416935 : Legacy Showroom
    0.407983 : LE MARCHÉ
    0.397825 : ALIEXPRESS STREET WEAR
    0.387168 : СНИКЕРХЭД | SNEAKERHEAD
    0.380457 : Кежуальные телки
    0.380159 : JAPANISM


    ['метро', 'utm_source', 'стоимость', 'анон', 'парк', 'utm_medium', 'районе', 'метров', 'москвы', 'utm_campaign', 'цена', 'площадь', 'улице', 'квартира', 'путешествие', 'кафе', 'планерной', 'квартиру', 'срочно', 'музей']
    0.998702 : Горящие туры - Москва
    0.994516 : Аренда Красноярск СОБСТВЕННИК
    0.975297 : Аренда квартир и вилл в Анталии
    0.974926 : SCARLETT JOHANSSON | СКАРЛЕТТ ЙОХАНССОН
    0.972857 : Сдать | Снять квартиру в Санкт-Петербурге
    0.971282 : Вторая очередь ДСЛ МГУ
    0.965415 : Снять квартиру комнату без посредников Стопагент
    0.963251 : Жилье в Москве без посредников Аренда Сдам Снять
    0.95593 : Локалс Москва - снять/сдать квартиру и комнату
    0.953498 : Свой уютный уголок Москва • Аренда квартир Снять
    0.947645 : wonderland | travel
    0.945694 : Москва Rent | Аренда Квартир
    0.945033 : Backpacker - Путешествия
    0.941497 : Привет, сейчас...
    0.941212 : ЖИЛЬЕ в Москве БЕЗ ПОСРЕДНИКОВ Аренда Сдам Снять
    0.939568 : Снять квартиру, комнату в Москве БЕЗ ПОСРЕДНИКОВ
    0.936252 : HORIZON
    0.933928 : Из окна вагона
    0.931353 : Туристическое агентство «Апельсин Travel»
    0.930959 : Bon Voyage | Идеи путешествий


    ['системы', 'воды', 'кожи', 'следует', 'организма', 'крови', 'тела', 'ученые', 'организм', 'кожу', 'мозга', 'животных', 'исследования', 'позволяет', 'уровень', 'эффект', 'веществ', 'являются', 'сортировку', 'ракетном']
    0.964926 : Вязание схемы
    0.910015 : Вязание со схемами
    0.908954 : Медицинская Фракция ARMADA
    0.894614 : Анатомичка
    0.880192 : Устройство авто
    0.87996 : Авто
    0.878076 : ГАССУЛ.РФ - Марокканская глина Гассул!
    0.876793 : Косметолог Орлова Анастасия. Балашиха
    0.874543 : Ветеринарная энциклопедия
    0.869972 : АвтоМастера BY
    0.866925 : Вязание крючком и спицами (Своими руками)
    0.860193 : Клуб "Вязание и рукоделие"
    0.859036 : Борьба с лженаукой в ветеринарии
    0.854524 : AGARU Cosmetics - натуральная косметика
    0.85238 : Анатомия человека
    0.846835 : Своими руками
    0.840613 : Вязаные идеи
    0.835811 : ОкейДок
    0.834493 : Дачник
    0.831988 : Идеи


    ['ингредиенты', 'масло', 'добавить', 'соль', 'приготовление', 'тесто', 'яйца', 'вкусу', 'перец', 'масла', 'сахар', 'салат', 'ложки', 'приятного', 'нарезать', 'сахара', 'мясо', 'аппетита', 'перемешать', 'молоко']
    0.995588 : Шедевры кулинарии | Вкусные рецепты
    0.995229 : Рецепты, Еда
    0.994855 : Кулинария-рецепты и советы
    0.994362 : COOKING TIME - время готовить!
    0.994192 : Вкусные рецепты любящей жены 💗
    0.993999 : НЯМКА. Кулинарные рецепты.
    0.991584 : Легкие рецепты
    0.991546 : Творчество на кухне | Интересные рецепты
    0.99119 : Вкусно!| Рецепты. Салаты. Выпечка. Десерты
    0.991103 : Рецепты с повышенным содержанием белка.
    0.990681 : Рецепты | Салаты
    0.990436 : Кулинарные Хитрости l Рецепты
    0.98968 : Tasty every day | Лучшие рецепты
    0.988991 : САЛАТЫ | Лучшие рецепты
    0.98878 : Записки шеф-повара
    0.988538 : ●ВКУСНЯШКА●Кулинарные рецепты.
    0.98796 : YUMMY | Рецепты 2018
    0.987949 : Bon Appétit | Лучшие рецепты
    0.987914 : Секреты кулинарии | Простые рецепты
    0.98785 : Рататуй. Кулинарные рецепты.


    ['суприме', 'вариция', 'instagram', 'новыйгод', 'myinspo', 'вдохновение', 'handmade', 'allvin', 'georgia', 'inst', 'rey_xd', 'reddit', 'essity', 'дизайнинтерьера', 'декор', 'style', 'сотовые', 'свадьба', 'infiniti', 'чтoбы']
    0.998104 : Megalyrics.ru
    0.988393 : Реальность? Нет, мне аниме по душе.
    0.916209 : Миллион музыки  ♪
    0.915427 : @INSTAGIRLS
    0.914873 : Квантовая Психология
    0.913946 : FAP INSTA
    0.903058 : Китайский язык | Учим по карточкам
    0.899002 : W.C.M.M
    0.897168 : Instagram photo | Инстаграм фото
    0.879675 : P E O P L E
    0.876849 : How to...
    0.875948 : Дневник Богини | БОТ
    0.867192 : Дурь ▶
    0.864735 : Юморист
    0.862576 : Я - САМАЯ... | БОТ
    0.851662 : Немагия  YouTube Ютуб
    0.829253 : DEEP HOUSE MAFIA  [D.H.M.]
    0.827866 : Это жизнь, детка, будь выше
    0.82646 : Girl`s Look
    0.82198 : Toyota CROWN. Japan style


    ['студентов', 'бизнес', 'курс', 'развития', 'программы', 'студенты', 'регистрация', 'обучения', 'университета', 'курса', 'обучение', 'компаний', 'проектов', 'управления', 'подробнее', 'конференции', 'образования', 'данных', 'сфере', 'российской']
    0.996107 : Колледж МТИ – отзывы
    0.995879 : Pepeliaev Group: Internship program
    0.994751 : Институт повышения квалификации специалистов
    0.99252 : Повышение квалификации   по ПТМ в Красноярске
    0.992442 : Институт Непрерывного Образования НОУ ВПО "ТИСБИ
    0.992235 : МТИ "ВТУ" г. Краснодар
    0.992213 : Повышение квалификации охрана труда. Красноярск
    0.991949 : Повышение квалификации по ПБ Новосибирск
    0.99181 : Дистанционные курсы повышения квалификации
    0.99181 : Курсы повышения квалификации 2018
    0.99181 : Курсы повышения квалификации для педагогов
    0.99181 : Курсы повышения квалификации для учителей
    0.99181 : Курсы повышения квалификации для воспитателей
    0.99181 : Курсы повышения квалификации
    0.991739 : Московский Технологический институт
    0.991667 : Повышение квалификации   по ПТМ в Новосибирске
    0.991518 : Повышение квалификации охране труда. Новосибирск
    0.991518 : Повышение квалификации Радиационная безопасность
    0.991441 : Повышение квалификации реставрации в Новосибирск
    0.991284 : Повышение квалификации эко. безопасность


    ['time', 'english', 'like', 'make', 'языка', 'students', 'запрашивая', 'язык', 'английский', 'take', 'work', 'people', 'английском', 'good', 'языке', 'life', 'know', 'year', 'английского', 'first']
    0.95775 : Фразы на испанском
    0.922688 : Электронные сигареты Черновцы l Vape Bar Cv
    0.898809 : Голландский (нидерландский) язык
    0.871444 : Репетитор английского в Зеленограде!
    0.866977 : Series English - English with TV series
    0.864286 : Английский язык в картинках
    0.861462 : GRE Free eBook and Practice Tests
    0.851714 : IELTS Scores
    0.84613 : Idiom Land
    0.841836 : IELTS Portal
    0.834815 : Канал Диалект | Простой Английский
    0.822517 : English Books Daily
    0.813453 : #Crossfit Music 2015 vol.1
    0.805792 : Английский — ExpertEnglish.Info
    0.79896 : Study Abroad: TOEFL/IELTS
    0.796026 : Словарный запас (EN)
    0.784708 : Amazing Girls
    0.770314 : Daily English: английский по фильмам
    0.760668 : Бесплатный Сочи
    0.757922 : Полидол | [/pol/ /hi/]


    ['суприма', 'цена', 'доставка', 'размер', 'скидка', 'скидки', 'заказать', 'платье', 'ссылка', 'заказ', 'водонепроницаемый', 'скейтеров', 'форумам', 'шрифте', 'дропнется', 'стёганый', 'арабик', 'majestic', 'наличии', 'продам']
    0.992025 : Канцелярия на AliExpress для ЛД, планеров
    0.99181 : Aliexpress style
    0.989205 : CANDYSHOP
    0.988881 : minimal store
    0.988444 : Товары до 1 Aliexpress
    0.98813 : Топ   Aliexpress
    0.97908 : Хайповый Алик
    0.976258 : Girl's AliExpress
    0.973706 : 776 Discount store
    0.971952 : LOOKS ALI
    0.969355 : Я есть хайп
    0.962524 : Simpson Store | анорак, парку
    0.95038 : Vintage Goods
    0.949563 : fashion aliexpress
    0.947286 : F S H N L K
    0.945856 : ГОДНОТА С АЛИЭКСПРЕСС
    0.943625 : ALIEXPRESS STYLIST
    0.941782 : НИШТЯКИ С КИТАЯ до 10
    0.935051 : Aliexpress Товары до 100 рублей
    0.9333 : Гель-лак. Токаревка.


    ['ворк', 'коллабы', 'супримом', 'вроде', 'овер', 'топов', 'гонзалеса', 'увидел', 'голову', 'общем', 'вдогонку', 'уйма', 'анархия', 'скриншоте', 'колю', 'драконами', 'ладно', 'битья', 'bonjour', 'компасом']
    0.974 : ИнстаХлам
    0.589897 : Вань, напиши мни :3
    0.548037 : NIGGALODEON
    0.5125 : ***ИЭП***
    0.465385 : Just photo
    0.41 : Доска объявлений  Г.Сибай
    0.403152 : Бот Повар | GIF рецепты
    0.399789 : Подслушанно ММУ №5
    0.364954 : Ты та самая нимфа
    0.338889 : МГУПС (МИИТ) ИЭФ. ЭМН-116.
    0.313528 : Ладно-Хорошо
    0.249285 : ☭БаНдА СГ☭СЕКТОР ГАЗА(wannabe band)ПЛЕМЯННИК ХОЯ
    0.229124 : соционика чота говорит


    ['коллабе', 'фильм', 'hanes', 'матч', 'сезона', 'жанр', 'команды', 'фильма', 'сезон', 'news', 'кино', 'серия', 'фадеев', 'madrid', 'hala', 'мемасиков', 'перевеса', 'серии', 'закупайтесь', 'йоркски']
    0.990209 : Я болею за «Барселону»!
    0.983898 : Футбольные Трансферы
    0.978969 : Прогнозы | ставки на спорт | мемы
    0.97498 : ТРАНСФЕРЫ
    0.974237 : Ставки/прогнозы/советы/жб/раскрутка---&gt;Victory
    0.974121 : Футбол | Ставки | Прогнозы
    0.971385 : BESTGOAL – ФУТБОЛ, ТРАНСФЕРЫ
    0.969731 : Английская Премьер-Лига
    0.9687 : Расписание Футбольных Матчей
    0.966147 : ФУТБОЛ: ТРАНСФЕРЫ
    0.962358 : TVGOAL - ФУТБОЛЬНЫЕ ОБЗОРЫ И ТРАНСЛЯЦИИ
    0.958696 : Футбольные Мемы
    0.958239 : Лига чемпионов УЕФА
    0.956997 : КАЗАНЬ | KZN TIMES
    0.954698 : Футбол – больше, чем игра!
    0.951875 : Titan-Bet
    0.950114 : Футбольные трансферы на Sports.ru
    0.948932 : ФУТБОЛЬНЫЕ ТРАНСФЕРЫ LIVE
    0.945539 : GOALTIME – ФУТБОЛЬНЫЕ ОБЗОРЫ И НОВОСТИ ФУТБОЛА
    0.945498 : Футбол в стиле Ретро (16+)


    ['альбом', 'коллаба', 'коллабов', 'music', 'альбома', 'фэйк', 'коллабами', 'музыка', 'трек', 'remix', 'black', 'билеты', 'feat', 'музыки', 'steep', 'anarchy', 'релиз', 'album', 'песни', 'love']
    0.996791 : Effloresce
    0.991533 : Rock | Metal
    0.991215 : Berghain Berlin
    0.987143 : _ripyourchest_
    0.98136 : Santa Muerte
    0.979787 : Music Life
    0.978222 : Λ R C O B A L E N O
    0.976462 : HDB
    0.97606 : ALDN
    0.975977 : Just... Breathe music
    0.973488 : E: Electronic music | Электронная музыка
    0.972909 : KILLY
    0.97229 : Diggin' Music
    0.970224 : DOPE
    0.968407 : Music Everyday
    0.966498 : Σ.G.O.
    0.965273 : engawa
    0.965116 : Vevo  Russia
    0.963462 : A10010111
    0.962782 : Технология


    ['бизнес', 'цели', 'бизнеса', 'будете', 'цель', 'клиентов', 'правило', 'идея', 'способ', 'результат', 'идеи', 'успеха', 'ситуации', 'решения', 'уровень', 'заниматься', 'книга', 'читать', 'план', 'процесс']
    0.982407 : tumblr salad shop/шмот на алиэкспресс
    0.952901 : Выйди из контакта и иди читать
    0.9525 : R E P L I C A
    0.943659 : Советские плакаты
    0.923888 : Бизнес клуб
    0.92348 : ДОСКА ОБЪЯВЛЕНИЙ | ОБЪЯВЛЕНИЯ В САНКТ-ПЕТЕРБУРГЕ
    0.897372 : Твой потенциал!
    0.895309 : Кружка кофе
    0.889844 : Финансовая грамотность
    0.88353 : Бизнес UP
    0.882386 : Записки Успешного Человека
    0.88125 : your station.
    0.876027 : Стив Джобс и другие великие бизнесмены
    0.875443 : Человек, который изменил всё
    0.872558 : Business Vision – бизнес-идеи, тренинги, финансы
    0.872347 : Маркетинговое Агентство "Группа-АБ"
    0.867421 : Российская академия автомобильного бизнеса.
    0.864286 : Zeitgeist 00
    0.864155 : СЛОВАРЬ
    0.860073 : Тренинги


    ['женщина', 'женщины', 'однажды', 'смерти', 'словно', 'женщин', 'тело', 'бога', 'мужчина', 'мужчины', 'смерть', 'сердца', 'чувства', 'отношения', 'земле', 'солнце', 'душа', 'отец', 'любить', 'мать']
    0.960334 : Дхаммапада
    0.959724 : Книга Книг
    0.949297 : Православие
    0.942196 : Иисус Христос • Учение Иисуса • Христианство
    0.941592 : † Архимандрит Иоанн Крестьянкин
    0.93311 : friedrich hölderlin (фридрих гёльдерлин)
    0.930068 : Митрополит Сурожский Антоний
    0.923846 : Великие люди о женщинах
    0.904077 : MUSIC BORDEL
    0.901084 : Мы будем утешены | ن
    0.894605 : Доброе слово •Православие•
    0.894568 : Хлеб и Зрелища
    0.891471 : Иеромонах Василий (Росляков)
    0.88636 : Архимандрит Кирилл (Павлов)
    0.885253 : † Православная мудрость
    0.884662 : One's inspiration
    0.88125 : delete
    0.880997 : Христианство ♰
    0.875766 : Уильям Шекспир
    0.873895 : Учение Будды


    ['меша', 'cheetah', 'поборник', 'deutsch', 'sich', 'nicht', 'superfigurka', 'club156801223', 'nooraland', 'emfas', 'avtoline', 'немецкий', 'немецком', 'vscofilters', 'splwars', 'childhoodruined', 'sonata', 'kurs', 'sein', 'eine']
    0.990777 : BMW Addict
    0.975089 : 👑Аваға арналған суреттер  👑
    0.967793 : QAZAQ PIE
    0.963937 : Похуғо маған ♕
    0.962295 : Жүрегім
    0.961135 : Даналық сөздер
    0.956553 : Қазақша хит әуендер |Жаңа әндер |Жана андер
    0.95354 : Немецкий язык для начинающих и знатоков
    0.952787 : A M Ó R ❣
    0.952291 : Бәрі бандит!Біз ғана простой☝♕
    0.950634 : Күлкістан ♕
    0.948969 : I N S
    0.940683 : Сен үшін, Братан.♕
    0.933693 : 💮 ПСИХОЛОГИЯ ❄ УЗБЕКСКИЙ-СТРАНА
    0.932143 : Психология жизни
    0.926923 : f r e s h
    0.925204 : ***AЛЛАҺ ЯРДӘМЕННӘН ТАШЛАМАС!***
    0.920833 : ХТКТК "КӨЗҚАРАС" ЗПК/ ХГТК ИДК "КӨЗҚАРАС"
    0.91713 : Нарисовал я пару штрихов
    0.913096 : AkidzabrA


    ['аллах', 'аллаха', 'россия', 'войны', 'власти', 'news', 'ссср', 'территории', 'республики', 'президента', 'стране', 'государства', 'путин', 'президент', 'населения', 'армии', 'российской', 'украины', 'заявил', 'глава']
    0.990594 : Камень в лесу
    0.936812 : Armenian Military Portal
    0.86016 : ИА Красная Весна
    0.856903 : Аль Интисор
    0.840726 : Ураза | Священный месяц Рамадан | Все о посте
    0.838996 : Сагъышла
    0.834085 : Кавполит
    0.830419 : Защищать Россию
    0.823502 : ® assalamu ale'kum
    0.820598 : Антимайдан-коммент
    0.812566 : Свободная Пресса
    0.812291 : КМД | Королевство Милых Дам
    0.810087 : Вайнаьх дийцар
    0.81 : Международный юридический институт
    0.807137 : МЕХКАĠONÇA
    0.805966 : Benelux military: Бельгия,Нидерланды,Люксембург
    0.801179 : INGINFORM l Обозреватель Ингушетии
    0.799345 : НОВОРОССИЯ | SaveDonbassPeople | Антимайдан
    0.79523 : АнтиБандеры | ДНР, ЛНР
    0.794186 : ВАЙ LAND


    ['пулловер', 'yankees', 'подсолнухами', 'маэда', 'киселеву', 'тренировки', 'мышцы', 'фотограф', 'упражнения', 'мышц', 'ноги', 'рост', 'упражнение', 'возраст', 'тренировок', 'мохаммеда', 'осваивайте', 'водолазок', 'коронная', 'толкайте']
    0.977688 : Женские ПОПКИ
    0.950289 : Свадебный фотограф Юлия Мусина
    0.897353 : Кастинги.ру
    0.889776 : Fitness &amp; Bodybuilding
    0.889416 : Кастинги.com
    0.879801 : Just do it | Stretching
    0.845057 : Кастинги в Москве
    0.817059 : IRON BODY - спорт
    0.796586 : Podium.Life - вакансии и кастинги для моделей
    0.790497 : yasafronova photography
    0.770273 : Кастинги съемки работа модели подработка
    0.759156 : ARTmosfera
    0.74873 : УТРЕННЯЯ ЗАРЯДКА!
    0.740043 : Массовка в Москве
    0.736374 : Промоутеры Москва + Event
    0.73078 : Рельеф —  сушка / фитнес / спорт / питание
    0.727899 : PLATYA V PROKATE | прокат платьев Москва
    0.725003 : Фотограф.
    0.724328 : 🎀 Photomodel 📸 Кастинги и фотосессии в Москве
    0.722517 : Кастинги www.ACMODASI.ru


    ['мама', 'знаешь', 'тобой', 'твои', 'девушка', 'говно', 'папа', 'родители', 'твой', 'парень', 'домой', 'напиши', 'можешь', 'хотелось', 'надеюсь', 'доме', 'спать', 'вроде', 'давай', 'понимаю']
    0.998119 : грусть как стиль жизни
    0.997003 : Фиолетовая папка
    0.996823 : Рядом.
    0.995697 : Стрельцы • 18+
    0.994994 : Ilita!
    0.983646 : ЩЕБЕСТАН
    0.983621 : Kalash.
    0.982407 : Пьяный Единорог
    0.981731 : HIGH SCHOOL
    0.981731 : Юмор
    0.981373 : Безысходность
    0.977229 : пьяна без вина.
    0.97625 : Жесткие Gif
    0.97625 : Подростки
    0.975 : D E A T H
    0.975 : Чайный листик 🍃 | Психология
    0.975 : Одиночество
    0.97352 : Девы • 18+
    0.972857 : Музыка Души
    0.972059 : Близнецы


    ['века', 'книга', 'искусства', 'фильм', 'искусство', 'культуры', 'wall', 'художник', 'книге', 'язык', 'книг', 'книгу', 'кино', 'философии', 'общества', 'русский', 'текст', 'автора', 'роман', 'художника']
    0.961626 : Аниме Зависимость | アニメ依存性
    0.961519 : Re FORMATICA (18+)
    0.905 : высота
    0.873871 : Религиоведение
    0.84932 : Книжная лавка
    0.838795 : level
    0.830595 : Подольский Институт Экономики (1 курс 2014)
    0.818403 : ФИЛОСОФИЯ | НАУКА | КУЛЬТУРА
    0.79694 : Сечение
    0.781481 : Тематический простор
    0.7701 : Кононспект искусствоведа
    0.769757 : Здешние песни
    0.768277 : ИСТОРИЯ ИСКУССТВ
    0.750095 : Египетское масонство в России
    0.741207 : философский штурм
    0.737675 : Театр. Книги.
    0.736688 : Мартинизм
    0.729788 : Арктогея
    0.727477 : О
    0.727094 : Книжная лавка «Алый Единорог»


    ['youtube', 'youtu', 'instagram', 'iphone', 'apple', 'watch', 'вконтакте', 'канал', 'twitch', 'самураями', 'вконтач', 'antihero', 'выбраный', 'сверхлёгкого', 'бунтарском', 'божескими', 'успешностью', 'дизайн', 'приложение', 'google']
    0.993613 : Обои для Apple iPhone и Android | Wallpapers iOS
    0.981771 : ПУТЕШЕСТВЕННИК
    0.975157 : Wylsa OA  Бесплатный Общий Аккаунт
    0.962381 : Royal Comics™
    0.934286 : Aoi
    0.914287 : The Best Anime Art - Лучшие Аниме Арты
    0.913636 : Лучший момент
    0.911384 : femfantasy
    0.892462 : Тайна Коко
    0.887873 : Game Whore | Cosplay x Art
    0.881195 : SEOUL
    0.869989 : Humanization | 擬人化 | Gijinka
    0.869468 : Волчонок | Teen Wolf | Скринкапсы
    0.839644 : Видеоархив BlackSilverUfa
    0.82502 : Steam Reviews
    0.822399 : ИГРОВОЙ ALIEXPRESS
    0.818162 : PLAY LIKE A GIRL
    0.815944 : Душа художника ( ART, АРТ )
    0.81043 : I ♥ 한국
    0.80917 : Queen of games
    
    
    
<img align="center" src="../images/3. publics/cat_o5.jpg" width="600"> 


# 8. Модель и 30 кластеров

Плохо мешать государство и религию, родителей и ванилек и многое другое. Попробуем увеличить число кластеров. Список из групп, вероятностей и слов снова закончится котейками. Внимательно посмотрите на всё это дело. 

    ['вход', 'билеты', 'мастер', 'ждем', 'москвы', 'концерт', 'адрес', 'санкт', 'приходите', 'праздник', 'стоимость', 'подарки', 'фестиваль', 'новогодние', 'метро', 'подробнее', 'сентября', 'фестиваля', 'состоится', 'билет']
    0.991369 : Relax Art ™
    0.981775 : S.O.V.A. - #ГЛАЗАОТКРОЙ
    0.918903 : Афиша Нижневартовска
    0.91234 : Свободное Пространство «ТИПИЧНЫЙ ПИТЕР» Антикафе
    0.906841 : у ю т н о с т ь
    0.888872 : КРЫША — ЛОФТ ПРОЕКТ ЭТАЖИ
    0.871425 : Каток в городском парке Орехово-Зуево
    0.867357 : FreeShows - бесплатные события Москвы
    0.861905 : черный файл
    0.861905 : Как такое возможно?
    0.858707 : КУРИЛЫ (ВДНХ)
    0.856328 : Мой Питер | Санкт-Петербург
    0.853784 : ТРЦ РИО "Кострома"
    0.852481 : Научное шоу ОТКРЫВАШКА в Санкт-Петербурге
    0.849566 : Фестивали Москвы
    0.84848 : Моя Москва
    0.841299 : Привет, сей час в Москве
    0.833027 : ГРК "Бережки-холл" Егорьевск
    0.830417 : ПАНДА ХУКА / PANDA HOOKAH [18+]
    0.828068 : Моя Москва | Moscow


    ['лого', 'коллаб', 'supreme', 'четверг', 'худи', 'бокс', 'расцветках', 'превью', 'футболки', 'вещей', 'куртку', 'расцветки', 'футболка', 'принтом', 'суприм', 'сезон', 'куртка', 'штаны', 'куртки', 'капюшоном']
    0.821029 : Хаос
    0.758333 : Lunnii kotik -- фотограф Аня Аксенова
    0.729855 : Jam's Stuff
    0.710698 : Журнал "Супримилка"
    0.607612 : Модный AliExpress/Поиск Товара
    0.59455 : Eat. tumblr. sleep
    0.58969 : AliExpress Хайповый Шмот
    0.571532 : AliExpress Хайповый Шмот
    0.561126 : F S H N L K
    0.535043 : THE PILE
    0.533207 : chanel's aliexpress
    0.516944 : Уличная мода. Фото со всего мира
    0.516667 : minimalism
    0.516667 : Twitter Dota
    0.513964 : Wish List
    0.503971 : LOOK HUNTER | В поисках идеального лука
    0.488502 : Годнота на AliExpress
    0.459417 : WOW SO ALI
    0.456801 : Legacy Showroom
    0.452049 : ALIEXPRESS STREET WEAR


    ['кисоньки', 'квартира', 'метро', 'евро', 'путешествие', 'стоимость', 'путешествия', 'отель', 'tattoo', 'квартиру', 'горы', 'квартире', 'комната', 'нура', 'travel', 'квартиры', 'цена', 'ночей', 'проживание', 'обратно']
    0.998679 : Горящие туры - Москва
    0.993803 : Backpacker - Путешествия
    0.985572 : MOST HANDSOME MAN
    0.973256 : Горячие туры и Горящие предложения!
    0.972381 : Сдать | Снять квартиру в Санкт-Петербурге
    0.960046 : Туристическое агентство «Апельсин Travel»
    0.955242 : Travel Spy | Дешевые билеты и путешествия
    0.952901 : SCARLETT JOHANSSON | СКАРЛЕТТ ЙОХАНССОН
    0.948418 : Аренда Красноярск СОБСТВЕННИК
    0.9373 : Аренда квартир и вилл в Анталии
    0.923142 : Путешествия
    0.913934 : Лето
    0.905533 : Страны мира
    0.902039 : Жилье в Москве без посредников Аренда Сдам Снять
    0.896658 : Турагентство выгодных туров "LAN TRAVEL"
    0.896457 : HORIZON
    0.892593 : Мгинский сервисный центр
    0.890716 : Дешевые туры, билеты, путешествия | LuckyWings
    0.885583 : Bon Voyage | Идеи путешествий
    0.883426 : Горящие туры в Турцию, Грецию, Сочи, на Кипр


    ['пулловер', 'мышцы', 'тренировки', 'стёганый', 'упражнения', 'тела', 'ноги', 'мышц', 'организм', 'организма', 'упражнение', 'следует', 'крови', 'воды', 'системы', 'тренировок', 'питания', 'уровень', 'пищи', 'упражнений']
    0.954801 : Вязание схемы
    0.939272 : Just do it | Stretching
    0.914081 : Рельеф —  сушка / фитнес / спорт / питание
    0.902535 : Вязание со схемами
    0.889094 : СПОРТ-ВЫБОР СИЛЬНЫХ / Бодибилдинг / Фитнес / ЗОЖ
    0.88554 : Любимая Качалка
    0.883131 : Sport Factor
    0.881212 : Программы тренировок для девушек
    0.879821 : СПОРТИВНАЯ МЕДИЦИНА. ОРТОПЕДИЯ. РЕАБИЛИТАЦИЯ
    0.877066 : Качалка Выручалка
    0.871227 : Easy fitness | Фитнес-гид. Просто о сложном.
    0.870402 : Fitness &amp; Bodybuilding
    0.86701 : BODY POLICE
    0.865928 : Медицинская Фракция ARMADA
    0.856541 : Угарная Качалка
    0.849011 : Спортивные статьи | Fitness
    0.840106 : Вязание крючком и спицами (Своими руками)
    0.839966 : 👊 NO PAIN NO GAIN 👊
    0.837265 : Crossfit [Кроссфит|программы тренировок]
    0.835634 : Упражнения на ноги фитнес как похудеть


    ['ингредиенты', 'масло', 'добавить', 'соль', 'приготовление', 'тесто', 'яйца', 'вкусу', 'перец', 'сахар', 'масла', 'салат', 'ложки', 'приятного', 'нарезать', 'аппетита', 'мясо', 'сахара', 'перемешать', 'яйцо']
    0.993101 : COOKING TIME - время готовить!
    0.993016 : Кулинария-рецепты и советы
    0.992435 : НЯМКА. Кулинарные рецепты.
    0.991444 : Рецепты, Еда
    0.99095 : Шедевры кулинарии | Вкусные рецепты
    0.990569 : Вкусные рецепты любящей жены 💗
    0.988177 : Легкие рецепты
    0.988027 : Кулинарные Хитрости l Рецепты
    0.987539 : Рецепты с повышенным содержанием белка.
    0.987205 : Творчество на кухне | Интересные рецепты
    0.987163 : Пальчики оближешь))) Самые вкусные и простые рец
    0.986579 : Вкусно!| Рецепты. Салаты. Выпечка. Десерты
    0.98507 : САЛАТЫ | Лучшие рецепты
    0.984786 : Рецепты
    0.984542 : ●ВКУСНЯШКА●Кулинарные рецепты.
    0.984529 : Tasty every day | Лучшие рецепты
    0.984093 : Секреты кулинарии | Простые рецепты
    0.983446 : Рататуй. Кулинарные рецепты.
    0.98328 : Just Cook
    0.983208 : Записки шеф-повара


    ['новыйгод', 'msdnru', 'myinspo', 'видное', 'club156801223', 'nooraland', 'emfas', 'handmade', 'allvin', 'vscofilters', 'mangaotaku', 'news', 'сургут', 'ленинскийрайон', 'miyagi', 'followme', 'fanfreitas', 'ручнаяработа', 'mbti4u', 'подписывайтесь']
    0.990615 : BMW Addict
    0.988294 : Реальность? Нет, мне аниме по душе.
    0.975214 : Бот Гороскоп
    0.938488 : I N S
    0.930952 : Психология жизни
    0.925641 : f r e s h
    0.924791 : 👑Аваға арналған суреттер  👑
    0.916975 : Нарисовал я пару штрихов
    0.896422 : DeepDream
    0.893014 : 💮 ПСИХОЛОГИЯ ❄ УЗБЕКСКИЙ-СТРАНА
    0.888257 : AkidzabrA
    0.882763 : Китайский язык | Учим по карточкам
    0.841277 : Toyota CROWN. Japan style
    0.807645 : w o n d e r l a n d
    0.798459 : Миллион музыки  ♪
    0.790117 : Референсы от УиТ
    0.783504 : Megalyrics.ru
    0.775012 : Типичный юрист
    0.766109 : Дурь ▶
    0.764346 : Instagram Girls


    ['бизнес', 'компаний', 'компания', 'рынка', 'бизнеса', 'рынке', 'млрд', 'банка', 'кейс', 'рынок', 'банк', 'пулом', 'скупке', 'эталонных', 'привлечённые', 'вконтач', 'редиректа', 'затёртой', 'business', 'роста']
    0.97013 : Повышение квалификации охрана труда. Красноярск
    0.965963 : Повышение квалификации эко. безопасность
    0.964944 : Повышение квалификации эко. безопасность Новосиб
    0.964558 : Повышение квалификации изыскателей в Новосибирск
    0.964159 : Повышение квалификации проектировщиков. Новосиб
    0.963793 : Повышение квалификации строителей в Новосибирске
    0.963325 : Повышение квалификации рабочих специальностей
    0.960615 : Повышение квалификации изыскателей в Красноярске
    0.95955 : Повышение квалификации   по ЖКХ в Красноярске
    0.959123 : Повышение квалификаци проектировщиков Красноярск
    0.957503 : Повышение квалификации НОП в Красноярске
    0.956011 : Повышение квалификации охране труда. Новосибирск
    0.950775 : Повышение квалификации строителей в Красноярске
    0.92624 : Повышение квалификации по ПБ в Красноярске
    0.908741 : Повышение квалификации   по ПТМ в Новосибирске
    0.906638 : Повышение квалификации энергоаудит. Новосибирскк
    0.906625 : Повышение квалификации Радиационная безопасность
    0.906209 : Повышение квалификации маркшейдеров. Красноярск
    0.904634 : Повышение квалификации Радиационная безопасность
    0.901994 : Повышение квалификации маркшейдеров. Новосибирск


    ['курс', 'google', 'вконтакте', 'html', 'онлайн', 'сайт', 'данных', 'сообщества', 'приложение', 'youtube', 'blog', 'дизайн', 'язык', 'wall', 'языка', 'брэндами', 'паблик', 'apple', 'android', 'telegram']
    0.971597 : O'Reilly Books
    0.891497 : VK API Change Log
    0.890514 : DevLog | веб, геймдев, программирование
    0.890322 : Подготовка к ЕГЭ от 60 баллов и выше. Школа ЕГЭ
    0.890278 : ЯПрограммист | Python
    0.888234 : Библиотека программиста
    0.887009 : //Я - Программист
    0.884939 : Программирование ITmozg:
    0.879451 : Программист
    0.879167 : EGE-GO - ОГЭ КДР ВПР Ответы (Запасной паблик)
    0.877922 : Ответы ГИА
    0.876629 : Репетитор английского в Зеленограде!
    0.865447 : Программирование
    0.861905 : Английский язык в картинках
    0.85968 : Обработчик формул Вконтакте
    0.858098 : Мембот
    0.851272 : ЯФ МГУПС (МИИТ) - Android
    0.84247 : Уютное сообщество программистов
    0.841152 : PyNSK - группа о Python
    0.838889 : R2D3


    ['цена', 'доставка', 'размер', 'instagram', 'заказать', 'заказ', 'платье', 'форумам', 'наличии', 'заказа', 'продам', 'скидка', 'магазин', 'состояние', 'скидки', 'ссылка', 'магазине', 'вопросам', 'цены', 'цвета']
    0.999286 : minimal store
    0.998155 : Simpson Store | анорак, парку
    0.997292 : Seeds
    0.990735 : Товары до 1 Aliexpress
    0.986555 : Иди сюда, я тебя нарисую
    0.983055 : Биржа рекламы Instagram | Реклама в инстаграм
    0.981693 : YIN-YANG shop
    0.981046 : W Ø E
    0.970033 : Aliexpress Товары до 100 рублей
    0.963664 : CANDYSHOP
    0.962803 : Товары с АliExpress
    0.962166 : Выгодный Aliexpress
    0.955725 : Одежда
    0.954277 : Биржа рекламы видеоаккаунтов Instagram
    0.953968 : G E R O Y ❄
    0.948726 : Светящиеся кроссовки
    0.94825 : PANSY SHOP
    0.940553 : 776 Discount store
    0.922522 : Фильтры Vsco
    0.918662 : Товары для художников из Китая


    ['сметут', 'фиксатором', 'пробитой', 'bonjour', 'водонепроницаемую', 'словно', 'солнце', 'небо', 'стихи', 'ветер', 'снег', 'сквозь', 'тобой', 'символике', 'вновь', 'халата', 'сбитого', 'покатаемся', 'тереть', 'ромб']
    0.996951 : Фиолетовая папка
    0.796397 : ..я б навеки пошёл за тобой
    0.758333 : Теория Абсурда
    0.758333 : я нет
    0.751872 : Давно не дети
    0.745222 : Nail place SPB
    0.724527 : makin' some magic
    0.722148 : ЛЕС
    0.715531 : Nik Арефьев
    0.6854 : Джио Россо
    0.677778 : я иногда боюсь что вы остываете когда спите
    0.677778 : ✪ATMOSPHERE✪
    0.655872 : Отрывки
    0.652029 : GreyMillenium
    0.641947 : PHOTOGRAPHER: Zotkina Anastasia
    0.639345 : р
    0.639323 : СТИХИЯ
    0.631805 : джек-с-фонарём
    0.630542 : КЛЮЧ
    0.628752 : Культурный арт-проект «Мужской голос»


    ['матч', 'сезона', 'фильм', 'команды', 'жанр', 'сезон', 'серия', 'madrid', 'фадеев', 'закупайтесь', 'матча', 'пятыми', 'hala', 'оправдавшая', 'калифа', 'команда', 'йоркски', 'серии', 'драма', 'лиги']
    0.983616 : Футбольные Трансферы
    0.981879 : Ilita!
    0.979002 : Ставки/прогнозы/советы/жб/раскрутка---&gt;Victory
    0.971194 : ТРАНСФЕРЫ
    0.968928 : ЩЕБЕСТАН
    0.966966 : ФУТБОЛ: ТРАНСФЕРЫ
    0.96295 : BESTGOAL – ФУТБОЛ, ТРАНСФЕРЫ
    0.959224 : Английская Премьер-Лига
    0.958427 : TVGOAL - ФУТБОЛЬНЫЕ ОБЗОРЫ И ТРАНСЛЯЦИИ
    0.95805 : Я болею за «Барселону»!
    0.951398 : GOALTIME – ФУТБОЛЬНЫЕ ОБЗОРЫ И НОВОСТИ ФУТБОЛА
    0.950719 : КАЗАНЬ | KZN TIMES
    0.949618 : Футбольные трансферы на Sports.ru
    0.944301 : EUROGOAL | Футбольные обзоры
    0.941025 : Футбольный мир
    0.939543 : Трансферы | Football News
    0.93757 : #ВсеНаФутбол
    0.935489 : Лига чемпионов УЕФА
    0.932784 : АПЛ
    0.932521 : Football News


    ['коллабов', 'коач', 'фэйк', 'нейлоновую', 'готик', 'vlone', 'yachty', 'экстравагантные', 'тэгов', 'доков', 'неиллюзорно', 'эпохальный', 'модника', 'херовое', 'выдержанно', 'миды', 'редкостями', 'woah', 'коллабом', 'дропать']
    0.83135 : MITSUBISHI - Автомобили Мицубиси
    0.79514 : Just Art
    0.776504 : Rock‖Рок
    0.758766 : Особенное настроение
    0.758333 : Это Же Наше Исκусс†во!
    0.754108 : zaebato
    0.736983 : Небесный Цеппелин
    0.663913 : ROFL 🎄
    0.66267 : Итальянский язык
    0.650471 : Avtosh Style
    0.649416 : 🚗ЗаАвто Москва продать/купить авто в МСК и МО
    0.645674 : Кальянная Friends (18+) - Ордынка
    0.578406 : Тачка за сотку
    0.57178 : BREAKDOWN SHOW
    0.561798 : Тачки за копейки
    0.558789 : MozartCitations
    0.549845 : My Home - Armenia
    0.547146 : Студия Автозвука и Тюнинга D Style Audio
    0.542383 : The Motrix | Mixing &amp; Mastering
    0.542111 : Авторынок Нижневартовск


    ['бизнес', 'бизнеса', 'цели', 'брэнды', 'клиентов', 'будете', 'цель', 'идея', 'правило', 'результат', 'идеи', 'успеха', 'брэндов', 'план', 'способ', 'процесс', 'решения', 'информацию', 'уровень', 'заниматься']
    0.982099 : tumblr salad shop/шмот на алиэкспресс
    0.914017 : Маркетинговое Агентство "Группа-АБ"
    0.906312 : Российская академия автомобильного бизнеса.
    0.90235 : Советские плакаты
    0.879167 : Акселератор онлайн-школ
    0.878012 : Wiki Business - бизнес, стартап
    0.861905 : Zeitgeist 00
    0.84548 : Кружка кофе
    0.833641 : РОСТ ВСЕМ! бизнес коучинг
    0.833159 : Бизнес клуб
    0.831434 : ДОСКА ОБЪЯВЛЕНИЙ | ОБЪЯВЛЕНИЯ В САНКТ-ПЕТЕРБУРГЕ
    0.823658 : Black Lion | Бизнес Журнал
    0.823408 : Клуб "Свое дело" – Идеи. Технологии. Тренинги.
    0.822456 : Механика бизнеса
    0.821828 : Стартап | Идеи для бизнеса
    0.821674 : Капитал | бизнес, финансы, инвестирование
    0.821071 : База поставщиков ВЕЩЕЙ
    0.814299 : Smart Money|Бизнес журнал
    0.813157 : Закон денег | Бизнес Успех Саморазвитие
    0.808363 : Идеи для Бизнеса


    ['женщина', 'женщины', 'отношения', 'любить', 'мужчина', 'тобой', 'однажды', 'чувство', 'мужчины', 'можешь', 'человеком', 'женщин', 'человеку', 'чувства', 'будь', 'тело', 'боль', 'твой', 'думать', 'читать']
    0.997999 : Омар Хайям и другие великие философы
    0.9973 : morfin.
    0.993958 : Реальность
    0.992835 : Факты о родившихся 28 мая
    0.991131 : Цепляет
    0.990583 : Как-то похер.
    0.989825 : Pyramid
    0.987226 : Пауло Коэльо и другие
    0.987173 : 5 умных мыслей
    0.986849 : FRIDAY.
    0.981402 : M Ø R P H I N Σ
    0.980588 : Статусы Со Смыслом
    0.979478 : Современная Философия
    0.978985 : вместе навсегда
    0.978775 : Yes
    0.976423 : Bossy
    0.975732 : Философия жизни.
    0.975214 : Fuera De La Ley
    0.975176 : × A T M O S P H Σ R Σ ×
    0.974912 : L'amour de Soi


    ['time', 'like', 'english', 'make', 'life', 'take', 'students', 'good', 'people', 'work', 'world', 'year', 'know', 'first', 'английском', 'look', 'love', 'would', 'английский', 'come']
    0.964936 : Hyacinth House
    0.962148 : IELTS Scores
    0.957248 : IELTS Portal
    0.956374 : Немецкий язык для начинающих и знатоков
    0.92126 : Deutschwelt - мир немецкого языка
    0.918302 : Illustration | Science
    0.900136 : ENGUROO ONLINE ENGLISH SCHOOL
    0.896728 : WORLD in English!
    0.892593 : Подготовка SAT и GRE
    0.888488 : English fabulous style.
    0.881837 : Ξnglish Discussions
    0.879167 :  DOGGO MEMES 
    0.873089 : ◉    #ingcirc
    0.871794 : Берешь и Рисуешь. Референсы.
    0.871673 : ЖИЗНЬ
    0.867711 : Edutopia
    0.865864 : Literatur auf Deutsch
    0.863566 : English Moscow Comedy
    0.852103 : English:Practice Makes Perfect
    0.851275 : O U T B U R S . T


    ['века', 'войны', 'музей', 'отстающий', 'клоунских', 'филлером', 'террасном', 'церкви', 'народа', 'ссср', 'храм', 'церковь', 'русской', 'армии', 'русского', 'народ', 'народов', 'русских', 'земли', 'название']
    0.971355 : Вторая очередь ДСЛ МГУ
    0.806667 : Международный юридический институт
    0.782078 : КМД | Королевство Милых Дам
    0.65907 : АРХИТЕКТУРА. ЧЕРТЕЖИ НА ЗАКАЗ.
    0.632844 : Europa Vetus
    0.627715 : Армяне | Армения
    0.627489 : Принесение мощей свт. Николая в Россию
    0.6175 : Кавказ сейчас
    0.612638 : Old Armenia
    0.606526 : Адыгэ макъамэ (Circassian music)
    0.599469 : Сказки для Бабы-ЕГИ - Кино. ЕГЭ. История
    0.593838 : Photo Gallery | Армения
    0.592817 : Колонны Казанского собора
    0.57958 : History Of Azerbaijan
    0.570968 : Кавказцы - герои России (блог Егора Бакаева)
    0.566929 : Native Dagestan | Дагестан
    0.562806 : Римская Империя
    0.561741 : "Къарачай бычакъ"-ножи ручной работы...
    0.560648 : Проекты домов. Строительство домов.
    0.556043 : Фото - история | Photo-History


    ['подсолнухами', 'оплата', 'возраст', 'девушки', 'фотограф', 'рост', 'съемки', 'переписками', 'ценивших', 'требуется', 'одежды', 'требования', 'график', 'макияж', 'запись', 'почту', 'требуются', 'близнецы', 'mail', 'модель']
    0.982042 : Кастинги.ру
    0.97082 : Кастинги.com
    0.959199 : Свадебный фотограф Юлия Мусина
    0.940699 : Кастинги в Москве
    0.927378 : LOFT | Лофт интерьеры
    0.901285 : Ivybush | Дизайн Интерьер Архитектура
    0.889765 : Кастинги съемки работа модели подработка
    0.879507 : Массовка в Москве
    0.875919 : Podium.Life - вакансии и кастинги для моделей
    0.868373 : 🎀 Photomodel 📸 Кастинги и фотосессии в Москве
    0.843927 : Кастинги www.ACMODASI.ru
    0.841529 : Кастинг, объявления, работа, массовка.
    0.837874 : Зрители на ТВ
    0.828749 : РАБОТА Промоутеры/Супервайзеры/Аниматоры/Курьеры
    0.822211 : ARTmosfera
    0.821532 : Art Zona! ВАКАНСИИ и РЕЗЮМЕ!
    0.818389 : Работа в Москве - BTL, Event, HoReCa, Съемки
    0.806667 : Beauty
    0.804052 : GET MODELS | Кастинги и работа в Event
    0.794116 : LUCSO MODELS MANAGEMENT Модельное Агентство СПб


    ['мама', 'говно', 'знаешь', 'папа', 'родители', 'домой', 'парень', 'начал', 'доме', 'твои', 'девушка', 'общем', 'пишет', 'вроде', 'нахуй', 'поняли', 'голова', 'живет', 'понял', 'паблика']
    0.987111 : ИнстаХлам
    0.982099 : Пьяный Единорог
    0.98141 : HIGH SCHOOL
    0.969792 : ВыЖеРодители
    0.961333 : Улётные приколы
    0.943137 : 4ch
    0.925641 : Live In Tattoo | Татуировки
    0.921495 : П Ё С Ь К А
    0.916357 : Безысходность
    0.912213 : ПОШЁЛ НАХУЙ КАЖДЫЙ ДЕНЬ.
    0.903333 : ᴠɪᴅᴇᴏsos
    0.903333 : Live In Art | Искусство
    0.903333 : Ответ по Фрейду
    0.894992 : 💥Фанфа про Святослава💥👼
    0.882566 : Упоротый Единорог
    0.879167 : ЗЛОЙ НЕГР
    0.858665 : FUN
    0.85245 : солидный pub`
    0.838889 : L A
    0.836606 : Четкие Приколы


    ['книга', 'века', 'фильм', 'искусства', 'роман', 'искусство', 'художник', 'книг', 'книге', 'книгу', 'смерти', 'кино', 'wall', 'автора', 'культуры', 'философии', 'фильма', 'картины', 'художника', 'теории']
    0.958263 : Аниме Зависимость | アニメ依存性
    0.947679 : Re FORMATICA (18+)
    0.930952 : Антология книжной эстетики
    0.903333 : высота
    0.837556 : ФИЛОСОФИЯ | НАУКА | КУЛЬТУРА
    0.807062 : Книжная лавка
    0.798488 : философский штурм
    0.793881 : Религиоведение
    0.793449 : Мартинизм
    0.773293 : PresentArt ●  Художественная студия в Москве
    0.771751 : Подольский Институт Экономики (1 курс 2014)
    0.760552 : Арктогея
    0.733279 : О
    0.724057 : Ролан Барт (Roland Barthes)
    0.723538 : Людвиг Андреас Фейербах
    0.72349 : Библиотека
    0.720425 : Esthesis
    0.716766 : Сочная литература
    0.715283 : Рене Генон
    0.707079 : иид


    ['iphone', 'животных', 'веган', 'knigivk', 'vegan', 'алания', 'apple', 'вегетарианство', 'веганство', 'thebassland', 'осетии', 'anime', 'naruto', 'childhoodruined', 'животные', 'tumblr', 'essity', 'осетия', 'queenofgames', 'сыроедение']
    0.953165 : ПУТЕШЕСТВЕННИК
    0.816679 : Aoi
    0.73208 : studyblr.
    0.730206 : Как я встретил вашу маму. Цитаты
    0.721105 : Ruined Childhood
    0.706068 : Head Rubbish
    0.703976 : РСО - АЛАНИЯ | ALANIA
    0.687948 : Humanization | 擬人化 | Gijinka
    0.647703 : "Живи Красиво - Умри Достойно"
    0.647434 : ♥  anime aesthetic ♥
    0.641975 : The Best Anime Art - Лучшие Аниме Арты
    0.617969 : Блич | Bleach
    0.60474 : Обои для Apple iPhone и Android | Wallpapers iOS
    0.592571 : Borshch 18+
    0.587858 : photo | iron version
    0.577407 : Обои для iPhone
    0.575563 : Queen of games
    0.566343 : Анатомичка
    0.554661 : Дни рождения аниме героев (5 января внутри)
    0.535042 : Хвост феи/Фейри тейл/ приколы  Fairytailmania :3


    ['коллаба', 'коле', 'ворк', 'слить', 'коллабы', 'супримом', 'овер', 'медведем', 'коллабе', 'гонзалеса', 'спустя', 'фильма', 'написал', 'войны', 'режиссёром', 'критиковать', 'уйма', 'наса', 'девяностые', 'анархия']
    0.659324 : Susanoo
    0.658187 : ОСТРОУМНЫЕ
    0.62998 : ТИПИЩНЫЙ ТАТАРИН [Official page]
    0.516667 : Студенческие истории
    0.508097 : Hisakura
    0.505556 : КОМИКСЫ
    0.484615 : ВЖУХ
    0.321663 : Типичный нарутовец| Наруто | Боруто | ТН(16+)
    0.270282 : Мемы HSE
    0.242196 : Фильмы HD
    0.207571 : Доктор Комаровский. Официальная страница
    0.195404 : Мужское кино
    0.187588 : Шедевры кино | Кино цитаты
    0.182091 : HD Фильмы | Новинки кино 2017


    ['кожи', 'волос', 'кожу', 'волосы', 'гусиным', 'братиков', 'плетёная', 'лица', 'ребенка', 'ребенок', 'средство', 'спать', 'масло', 'кожа', 'масла', 'маска', 'лицо', 'крем', 'малыша', 'средства']
    0.992177 : грусть как стиль жизни
    0.966667 : Строки пыльных книг
    0.935107 : Другая красота
    0.925641 : Оптимист
    0.922321 : CALIFORNIA MAGIC
    0.912121 : Душа под Маской
    0.90808 : Макияж
    0.90203 : all i need
    0.887232 : Цитаты и Книги
    0.87212 : Идеи Комфорта-дизайн интерьеров, мировые новости
    0.866055 : Странный юмор
    0.838889 : Собаки
    0.838442 : LOOKBOOK
    0.812125 : Маникюр | Дизайн ногтей
    0.809253 : Дневник Богини | БОТ
    0.806667 : Знатница
    0.806667 : Упоротый Twitter
    0.774566 : Прически | Стрижки 💇 Окрашивания
    0.766598 : Street Styler | Уличная мода
    0.762649 : Я - САМАЯ... | БОТ


    ['альбом', 'меша', 'music', 'коллабами', 'альбома', 'remix', 'музыка', 'feat', 'музыки', 'steep', 'clarks', 'black', 'album', 'трек', 'records', 'релиз', 'house', 'билеты', 'музыку', 'rock']
    0.998252 : Oval Sound
    0.986286 : Effloresce
    0.982103 : ALDN
    0.980624 : Santa Muerte
    0.979433 : Music Life
    0.979279 : Berghain Berlin
    0.975589 : Armin van Buuren © A State Of Trance
    0.971928 : TEAM 140
    0.967443 : WCM | World Club Music
    0.965029 : Metal Clips | Метал Клипы
    0.959478 : Armin van Buuren • Together in A State of Trance
    0.957082 : ОЛИМП
    0.955843 : Tommy Genesis
    0.954644 : Σ.G.O.
    0.953968 : Dope Sound
    0.952694 : Speeeedy EDM Blog
    0.951227 : Diggin' Music
    0.947719 : Hardwell
    0.94689 : Ultra Music Festival | UMF
    0.945405 : Old School Melodic Death Metal


    ['пробуждающие', 'крашер', 'вайтом', 'потухать', 'добавляю', 'весе', 'чемпион', 'хабиб', 'поединок', 'чемпиона', 'superfigurka', 'нурмагомедов', 'учим', 'бокса', 'конор', 'испанский', 'patatas', 'долларов', 'чемпионом', 'хабиба']
    0.999108 : Даналық сөздер
    0.991089 : Жүрегім
    0.987243 : QAZAQ PIE
    0.985445 : За Тебя Мама
    0.983947 : Похуғо маған ♕
    0.980057 : Бәрі бандит!Біз ғана простой☝♕
    0.978772 : Қазақша хит әуендер |Жаңа әндер |Жана андер
    0.978263 : Добавь в друзья
    0.97757 : Фразы на испанском
    0.971594 : Күлкістан ♕
    0.966275 : Сен үшін, Братан.♕
    0.960004 : Мүмкін емес, бірақ шындық ♕
    0.932398 : Жаңалықтар
    0.925403 : Даналық сөз
    0.919444 : ХТКТК "КӨЗҚАРАС" ЗПК/ ХГТК ИДК "КӨЗҚАРАС"
    0.91666 : Fc Astana...QazaQstaN
    0.880604 : Дербисевашки
    0.871721 : SHAMSI 👑
    0.86037 : Гиннес рекордтар кітабы!
    0.860079 : СШ "Икан"


    ['маэда', 'шрифте', 'instagram', 'технарям', 'лиственный', 'spitfire', 'липучками', 'хайпбист', 'джпегов', 'футурой', 'photo', 'фотограф', 'model', 'dance', 'фотостудия', 'анастасия', 'анна', 'photographer', 'танца', 'fashion']
    0.963818 : A M Ó R ❣
    0.956382 : AIESEC в Україні І Вперед на стажування
    0.948528 : Anastasiya Maltseva
    0.934292 : Галерея живых картинок
    0.933377 : @INSTAGIRLS
    0.928778 : Твій простір можливостей
    0.926986 : ФРІ Plus Львів
    0.924681 : Турклуб " Манівці "
    0.921991 : AIESEC у Львові
    0.917516 : ***AЛЛАҺ ЯРДӘМЕННӘН ТАШЛАМАС!***
    0.912121 : Women estetiq
    0.907623 : Институт непрерывного специального образования
    0.904577 : FAP INSTA
    0.904406 : Студентське братство Сумщини
    0.904246 : Форум видавців
    0.90001 : Instagram photo | Инстаграм фото
    0.899184 : КНУ. Факультатив з алгоритміки
    0.889099 : Файні меми про українську лiтературу
    0.887018 : Татарча кызыграк
    0.882879 : VK Україна


    ['студентов', 'студенты', 'состоится', 'регистрация', 'университета', 'программы', 'обучения', 'развития', 'российской', 'факультета', 'образования', 'ранхигс', 'мероприятия', 'школы', 'курса', 'участники', 'федерации', 'апреля', 'курс', 'института']
    0.991667 : Дистанционные курсы повышения квалификации
    0.991667 : Курсы повышения квалификации 2018
    0.991667 : Курсы повышения квалификации для педагогов
    0.991667 : Курсы повышения квалификации для учителей
    0.991667 : Курсы повышения квалификации для воспитателей
    0.991667 : Курсы повышения квалификации
    0.987446 : "Учебный центр повышения квалификации АПК РТ"
    0.982325 : Кадровый портал Ленинградской области
    0.975833 : Институт переподготовки и повышения квалификации
    0.961333 : ЮЖНО-УРАЛЬСКИЙ ИНСТИТУТ УПРАВЛЕНИЯ И ЭКОНОМИКИ
    0.951667 : Аттестация рабочих мест, повышение квалификации
    0.946301 : Кумертауский институт экономики и права
    0.946296 : Российская академия народного хозяйства
    0.937514 : Центр Непрерывного Образования (ЦНО)
    0.930956 : Московский Социально - Гуманитарный Институт
    0.927798 : Кадровый портал Администрации Санкт-Петербурга
    0.916051 : Институт государственной службы и управления орг
    0.91419 : Абитуриент УИУ РАНХиГС
    0.912121 : МГАУ им. В.П.Горячкина. Энергетический факультет
    0.90762 : Уральский институт экономики Санкт-Петербургског


    ['суприма', 'youtube', 'скриншоте', 'hanes', 'хоккейный', 'дропнется', 'youtu', 'twitch', 'watch', 'подписываетесь', 'напарников', 'стопудово', 'анонсу', 'шрифтами', 'рыночек', 'контрастностью', 'necessary', 'двухсторонние', 'фэйл', 'изменённым']
    0.984388 : Лайк Тайм
    0.961398 : Royal Comics™
    0.95474 : Забей, девочка. Ему похуй.
    0.933417 : Не пытайтесь повторить это дома
    0.922403 : По сути похуй †
    0.919444 : Indie
    0.908604 : Academy Dota 2
    0.894238 : Porsche Cayenne - Порше Кайен
    0.892593 : CS:GO Stuff 🎄🎅
    0.888461 : Голос ПЛЮС+ [Фотобатлы] [Оопросы]
    0.879167 : RETCON | Marvel/DC
    0.872191 : Dota Mania
    0.870522 : E U P H O R I A
    0.869334 : SGL
    0.868244 : Team Secret
    0.864148 : FRAG M0V1E  [CS 1.6 | CS:GO | WARFACE]
    0.856752 : История D
    0.845703 : MMA Breakdown Video
    0.838889 : sibNET.CSGO
    0.838119 : DEEP HOUSE MAFIA  [D.H.M.]


    ['анон', 'анонимно', 'utm_source', 'utm_medium', 'подскажите', 'utm_campaign', 'мемасиков', 'ребят', 'жопастенькая', 'слушайти', 'здравствуйте', 'добрый', 'админ', 'помогите', 'админу', 'девушку', 'срочно', 'заранее', 'прошу', 'social']
    0.948177 : Edebiyat &amp; Şiir (Турецкая литература и поэмы)
    0.919444 : мни данаму кожитса
    0.912121 : Подслушано Балашиха (запасная)
    0.890872 : Üstadlardan Güzel Sözler
    0.858905 : ∑xWave
    0.858668 : ДИПЛОМ КУРСОВАЯ РЕФЕРАТ ИУИИ ИУИ ИУИИ ИУИ
    0.858658 : ДИПЛОМ КУРСОВАЯ РЕФЕРАТ МИСБ МАТИФТ МАТФТ МВШЭ
    0.858651 : ДИПЛОМ КУРСОВАЯ РЕФЕРАТ ИУЭПИИ ИУЭПИ ИФТИИ ИФТИ
    0.858517 : ДИПЛОМ КУРСОВАЯ РЕФЕРАТ РАНХиГС МИО МИР МИЭИП
    0.858409 : ДИПЛОМ КУРСОВАЯ РЕФЕРАТ МИБИСС МИБСС МШСЭН
    0.838889 : Подслушано АСО (КСЮИ)
    0.838889 : dlmbln
    0.810406 : SURVIVOR 2017
    0.806667 : ЮЗБ-3(МНЭПУ)
    0.787578 : In Borderlands We Trust
    0.778644 : Селфи Молодечно
    0.777467 : КС:ГО &amp; PUBG
    0.758333 : Развратница, ты моя
    0.758333 : М ×
    0.750195 : O Ses Turkiye


    ['yankees', 'news', 'россия', 'власти', 'кларком', 'молодёжно', 'стране', 'html', 'государства', 'войны', 'ссср', 'ранее', 'территории', 'населения', 'источник', 'словам', 'российской', 'глава', 'президента', 'президент']
    0.990429 : Камень в лесу
    0.920201 : Коммерсантъ Власть
    0.894922 : Евгений Фёдоров :: официальная страница
    0.876376 : ИА Красная Весна
    0.871751 : Федеральное агентство новостей
    0.867851 : Нижневартовск
    0.86445 : Анатолий Шарий | Новости | Украина | Sharij.net
    0.864055 : Антимайдан-коммент
    0.862837 : Военное обозрение
    0.853772 : Защищать Россию
    0.852753 : НОД :: официальная группа
    0.849325 : Так себе факты
    0.835513 : РБК
    0.831829 : Новая газета
    0.828956 : ЧП [В] ХМАО.
    0.827471 : ТРЕТЬЯ МИРОВАЯ ВОЙНА
    0.826468 : НОВОРОССИЯ | SaveDonbassPeople | Антимайдан
    0.826142 : Телеканал ЦАРЬГРАД
    0.823494 : ОПЦН ЮГО-ВОСТОЧНАЯ ФЕДЕРАЦИЯ
    0.822467 : Политика без купюр


    ['аллах', 'аллаха', 'суприме', 'купля', 'шейх', 'пророк', 'посланник', 'имам', 'благословение', 'пророка', 'аллаху', 'всевышний', 'благословит', 'коран', 'الله', 'приветствует', 'господь', 'доволен', 'поистине', 'корана']
    0.899715 : Ислам
    0.887521 : Ислам | Полезные наставления|Muslim Islam |Акыда
    0.867976 : Ислам
    0.867919 : Аль Интисор
    0.845651 : Ураза | Священный месяц Рамадан | Все о посте
    0.828524 : «Ислам - это Коран и  Сунна
    0.823372 : Ата Джуртну ауазы
    0.808656 : Интересно об Исламе
    0.805108 : Намаз по Сунне
    0.800422 : ОБ ИСЛАМЕ
    0.793996 : Мой идеал - пророк Мухаммад ﷺ
    0.786029 : Пророческий Путь -  منهاج النبوة
    0.774885 : Голос Ислама | The Voiсe of ISLAM
    0.772229 : Призыв к Саляфии ИДЕЛЬ - УРАЛ
    0.770513 : RPG
    0.769694 : ВАЙ LAND
    0.76853 : Сагъышла
    0.756872 : Статьи по манхаджу
    0.754233 : Анти-суфизм | Заблуждения суфизма
    0.749 : ® assalamu ale'kum


<img align="center" src="../images/3. publics/3cat.jpg" width="800"> 


Как минимум, нам удалось отделить государство от религии. Также у нас появилось несколько разных кластеров, связанных с мемами. Возможно, это уже говорит о избытке кластеров. Ну либо мемы, действительно, разные. Мы сохраним результаты обеих кластеризаций. Они нам понадобятся в следующих частях рисёрча. 

# 9. Tf-idf

Нам уже удалось получить неплохую кластеризацию. Попробуем теперь на этом же корпусе текстов немного иной подход для их предобработки, который называется tf-idf. 

**Tf-idf** — это чуть более сложный подход к созданию из текстов матрицы признаков. Как и раньше мы будем считать, что если слово в тексте встречается часто и оно при этом не является стоп-словом, то оно важное. К этому мы добавим, что если слово встречается в других документах реже, чем в данном, то и в этом случае, скорее всего, оно важно для текста.
По этому слову можно отличить этот текст от остальных. Если учесть описанные выше соображения, в результате получится подход tf-idf.

Значение признака для слова $w$ вычисляется по формуле: 

$$ 
w_{i,j} = tf_{i,j} \cdot ln(\frac{N}{df_{i}})
$$ 

* $w_{i,j}$ - tf-idf вес для слова $i$ в документе $j$
* $tf_{i,j}$ - Количество упоминаний слова $i$ в документе $j$
* $df_i$ - Количество документов, которые содержат слово $i$ 
* $N$ - Число документов 


Если слово часто встречается в тексте, то оно важно для него. За такую важность отвечает множитель $tf$ (term frequency), значение которого при этом раскладе, оказывается высоким. Второй множитель называется $idf$ (inverse document frequency, обратная документная частота). Если слово редко встречается в других документах, но при этом встречается в нашем, значение $idf$ будет большим. Если слово встречается в каждом тексте, то значение признака будет нулевым $(\log l = 0)$. В итоге произведение этих двух метрик определяет важность каждого слова. 

Мера tf-idf часто используется в задачах анализа текстов и информационного поиска, например, как один из критериев релевантности документа поисковому запросу, при расчёте меры близости документов при кластеризации. Попробуем этот подход на изначальном корпусе и на корпусе без стоп-слов. 

Такой подход выдаст нам две плохие кластеризации. Практически все кластеры представляют из себя какую-то жесткую мешанину. 

```
['tvoykek', 'club122485279', 'хитростижизни', 'лейми', 'club151125407', 'пeсен', 'искaл', 'стрoки', 'адскаякухня', 'воть', 'blasterjaxx', 'hexagon', 'your_psycho_life', 'аееееееееееееееееееее', 'арцах', 'club156437731', 'егo', 'затянет', 'club32169030', 'club139431937']
0.748367 : Миллион музыки  ♪
0.730552 : Дурь ▶
0.693359 : станция №12
0.620832 : К О С Ь К А
0.535856 : чёрный фон
0.529916 : Правильная Музыка
0.525 : ≥RaidCall≤  ≥ 5953698 ≤
0.525 : Психология жизни
0.46187 : http://
0.458472 : The Motrix | Mixing &amp; Mastering
0.446459 : вкусные пикчи
0.421646 : Draw U!
0.407172 : Армения - Арцах ♰ Настоящий Армянин
0.394458 : Lé Cilicia | Կիլիկիոց
0.373636 : Хитрости жизни
0.364295 : SOUNDPARK DEEP
0.363871 : МЭГУ | Диплом, Курсовая, Реферат

['суприм', 'undercover', 'овер', 'сметут', 'кашемировый', 'madame', 'нейлон', 'аэродрома', 'чуваку', 'slayer', 'приступе', 'пионерами', 'воплей', 'шнурках', 'скейтер', 'липучках', 'крикливая', 'leather', 'сформировало', 'intended']
0.693604 : Literatur auf Deutsch
0.685029 : Melancholia
0.667525 : Deutschwelt - мир немецкого языка
0.570355 : Академия Смешных Людей 👑
0.570253 : Фильмы на немецком языке / Filme auf Deutsch
0.567426 : Deutsch ist GEIL!!!
0.566125 : In Borderlands We Trust
0.562998 : TestDaf / Test DaF / DSH und andere Prüfungen
0.545248 : B Ʌ B
0.535599 : MusicOxygen
0.530773 : Узнай Всё
0.520788 : Дореволюціонный Совѣтчикъ
0.520616 : Немецкий язык для начинающих и знатоков
0.520577 : Deep LoVe Show
0.514819 : Olympus mju II μ[mju:]-II
0.508204 : mini tattoo
0.507732 : Party bus Surgut
0.504089 : TATTOO MASTERS
```

В голову как-то не очень приходят слова, которыми можно всё это описать, как это было в случае предыдущих двух моделей.  Забудем об этих моделях...

# 10. Генерируем новый контент

Бывают модели, с помощью которых можно генерировать какие-то новые объекты: видео, музыку, тексты. Обычно этим занимаются нейронные сетки, о которых мы поговорим немного позже. Например, одной из самых знаменитых генераций является альбом [группы нейронная оборона.](https://meduza.io/shapito/2016/06/28/neyronnaya-oborona-robot-napisal-teksty-pesen-pod-egora-letova)  Все тексты песен были сгенерированы нейросеткой, обученой на стихах Егора Летова. Мне больше всего нравится песня [заново мир](https://music.yandex.ru/artist/4445922?from=serp)

Как мы уже говорили выше, матричное разложение, использующееся в LDA, имеет вероятностную интерпретацию. Это означает, что с помощью LDA можно генерировать новые тексты. Если обучить модель на большом корпусе из ниров, можно будет заставить её написать за вас нир. Скорее всего, он выйдет не очень качественным. Качественные генерации делают нейросетки. Уже было [несколько скандалов](https://habrahabr.ru/post/213891/) связанных с тем, что какие-то чуваки нагенерировали ненастоящих статей, а журналы их опубликовали. Например, в России [в публикацию попала статья](http://polit.ru/article/2009/04/01/erunda/) «Корчеватель: алгоритм типичной унификации точек доступа и избыточности». Вроде бы её название звучит, как нечто ультраглубокое и годное... 

Итак, процесс порождения текстов описывается следующим образом:

Для документа $d$ длины $n_d$:
1. Из распределения Дирихле с параметром alpha сгенерировать распределение над множеством тем: $\theta_d \sim Dirichlet(\alpha)$
1. Для каждого слова $w = 1, \dots, n_d$:
    1. Сгенерировать тему из дискретного распределения $t \sim \theta_{d}$
    1. Сгенерировать слово из дискретного распределения $w \sim \phi_{t}$.
    
Подробнее об этом в [Википедии](https://en.wikipedia.org/wiki/Latent_Dirichlet_allocation).

В контексте нашей задачи получается, что, используя данный генеративный процесс, можно создавать новые группы и новый контент для них. Например, с помошьб модели `ldamodel_30` можно сгенерировать следующий текст из 30 слов: 

```
обстановка
свободный
порадуйте
сергей
девочки
наслаждаемся
веселиться
разным
книгой
руслана
ждем
настоящая
конкурс
фехтованию
виталий
поиске
октябрьская
забег
размеров
представляем
```

Неплохая вышла бы группа. Я бы подписался. На этом остановимся в наших упражнениях по кластеризации и двинемся дальше. 

<img align="center" src="https://raw.githubusercontent.com/FUlyankin/Parsers/master/images%20/cats.jpg" height="1200" width="1200">
