## 1 О том, что остаётся в интернете

Одним прекрасным утром я проснулся у себя дома и пошёл на кухню, чтобы заварить себе чай и взять пару пироженок. Попивая чаёк, я планировал просматривать мемы и читать вконтакте иканамовский чатик. Утро было спокойным и безмятежным. Ничего не предвещало беды. Когда я открыл чатик, я увидел как пацаны над чем-то угарают. "Должно быть новый мемас",- подумал я. Это был не мемас. Промотав чатик чуток повыше, я увидел нечто из прошлого. И, конечно же, орнул. В чатике лежал следующий скриншот:

В 2009 году, когда я учился в 8 классе, я жёстко угарал по защите окружающей среды. У меня даже была группа вконтакте под названием "Мир" с диванной аналитикой по глобальному потеплению. В пиковый период своей зелёной деятельности, я нашёл на каком-то левом сайте проект с капсулами времени и запилил послание потомкам, которые без моей мудрости явно бы не выжили. В итоге, спустя 8 лет, интернет ни о чём не забыл, и Сёма Осипов каким-то образом откопал этот сайт. **Всё, что попало в интернет, навсегда остаётся в интернете.** Каждый из нас оставляет за собой какой-то информационный след, и такие эпичные записи далеко не самая ценная его часть. 

Наш след помогает всем заинтересованным понять какие мы с вами на самом деле: насколько открытые, умные, социализированные, что любим покушать, что хотели бы купить и т.п., а после как-то использовать всю эту информацию. Например, посмотреть [насколько часто похожие на нас люди](https://www.rbc.ru/technology_and_media/06/10/2017/59d6b4309a79475f4c3e5c59) возвращают кредиты и решать на основе этого давать нам его или нет.

Все [наши лица  — большие данные,](https://lifehacker.ru/2016/04/08/your-face-is-big-data/) и мы уже ничего не сможем поделать с этим. Даже если мы решим почистить за собой всё то, что наследили, это не особо поможет. В сети существует огромное количество различных [баз с данными,](https://socialdatahub.com/ru/sbor_i_strukturirovanie_dannyih) которые на ежедневной основе собирают всё то, что мы оставляем за собой. Более того, подобные базы есть у любой крупной компании. Всё это нужно, чтобы творить магию машинного обучения. 

Именно машинное обучение может помочь [магазину узнать о вашей беременности раньше, чем об этом узнает ваш отец](http://ru.datasides.com/магазин-узнал-о-беременности-девушки/) и даже ненароком сдать вас ему. Оно может помочь вам найти свой собственный [закон пива и подгузников](http://tmstudent.ru/article/keis-pampersy-i-pivo) и повысить прибыль. Более того, оно может помочь вам грамотно сагетировать избирателей и [одержать на выборах победу.](https://theins.ru/politika/38490) Вокруг последних выборов идут довольно жаркие споры. Однако, прочитав книгу [Верховный алгоритм](https://www.ozon.ru/context/detail/id/136780904/), можно узнать, что судьбу США в 2012 году определило именно машинное обучение. Обама оказался первопроходцем в использовании больших данных в политике. Кстати о политике. Кажется, [один парень](https://habrahabr.ru/company/fbk/blog/335778/) впечатлился всем этим. 

Иногда сбор данных о вас происходит очень неожиданным образом. Например, сегодня социальной сетью вконтакте владеет Mail.Ru Group. Не так давно он [начал эксперементировать с психотипами людей.](https://www.youtube.com/watch?time_continue=1&v=swnnpqu7a6s) В соц сети  было запущено приложение-тест, пройдя который, пользователь мог узнать немного больше о своей психологии. Попутно он рассказывал о своём психотипе мэйлу. Вся информация тщательно собиралась и в итоге, собрав достаточные объёмы данных, mail смог обучить модель, которая по соодержимому странчки человека прогнозирует его психотип. Сегодня эти данные используются для оценки работоспособности потенциальных сотрудников. Использует их не только мэйл. Вполне возможно, что ваше резюме будет отобрано или не отобрано из кипы других именно благодаря модели, обученой находить годных людей именно по этой выборке. 

На самом деле социальные сетки и поисковые запросы — не самое крутое, что можно найти. В последние годы наличные деньги уходят на задний план. Всё больше транзакций мы совершаем в одно касание с помощью банковских карт. Это приводит к тому, что банкам доступна детализированная информация о всех наших покупках. Узнать в красках о том, какие ещё информационные следы оставляют люди, можно в [видосе одного блогера на ютубе.](https://www.youtube.com/watch?v=qBDbO_bAdFM&t=21s) Рассказано красиво, в эмоцях и немного натянуто. 

На самом деле бояться нечего. Ну подумаешь, какой-нибудь банк купил информацию о том что ты гуглишь, и использует её при принятии решения выдавать тебе кредит или нет. Во-первых, база данных будет довольно большой. Во-вторых, она будет продана в анонимизированном виде. Какое наблюдение соответствует какой фамилии, будет знать только продавец. В третьих, поисковые запросы отдельного человека не очень сильно помогут при обучении модели. Данных нужно много. Да и единственное, что может произойти, это то, что аналитик поугарает над тем какое именно порно вы вчера искали, а потом,  дома, в режиме инкогнито, тоже попробует его посмотреть. Пока ты не нарушаешь закон, всем на тебя пофигу. А когда ты его нарушишь, тогда тебя, скорее всего, найдут. 

**Я рассказываю всё это не для того, чтобы напугать, а для того, чтобы показать возможности и промотивировать ботать.** В ближайшие годы [спрос на высококвалифицированную рабочую силу,](http://tass.ru/pmef-2016/article/3345977) к которой относятся и аналитики, будет расти. Спрос на специалистов со средней квалификацией (бухгалтера, юристы и тп) будет падать. [Многие профессии исчезнут.](https://vk.com/ppilif?z=video-75410445_456239067%2F3c06fddeb99f78371a%2Fpl_post_-75410445_3892) Если вы не сожете успевать бежать за временем, ваша работа, также как работа 3000 юристов в сбере, рано или поздно [будет передана роботу,](https://www.rbc.ru/rbcfreenews/5877b2979a79478752358fb9) которого обучили дата сайнтесты. Не поймите меня неправильно. Юристы не исчезнут, они останутся. Но только самые высококвалифицированные.  

Будущее наступает. И вы можете не просто жить в нём, а творить его. Сегодня в мире довольно сильно не хватает спецов по анализу данных. Хороший дата-сайнтист должен уметь прогать как минимум на двух языках, хорошо знать математику, разбираться в кипе алгоритмов, понимать какими бывают данные, как их можно добывать, обрабатывать и хранить. На то, чтобы всё это поместить в свою голову и как следует упорядочить, нужно убить уйму времени. Без должной мотивации ничего не выйдет. Поэтому главная цель моего рисёрча именно в этом: показать какие интересные штуки бывают на этом свете и замотивировать неистово ботать. 

## 2  О целях и задачах иканам гранд рисёрча 

Как мы уже выяснии, **главная цель:** замотивировать тебя ботать.  Изначально я хотел ограничиться сбором данных из приказов о зачислении и социальных сетей, а также последующим строительством на основе этих данных модели, которая предсказывала бы вероятность того, что человек сможет закончить первый курс. Наверное, знание такого прогноза, в какой-то степени, мотивирует ботать. 

Однако всё вышло из-под контроля. Сначала мне захотелось посмотреть на граф дружеских связей эконома и попытаться понять какими характеристиками он обладает. Потом мне захотелось посмотреть на то как курсируют лайки внутри эконома. Мне захотелось понять какие люди на экономе более авторитетны, какие люди генерируют годный контент, есть ли какие-то взаимные симпатии между людими. На самом деле я задал себе много подобных вопросов. Беда в том, что всё это скрыто от наших глаз, и судить мы можем об этом только по внешним проявлениям. Внешние проявлеия, в свою очередь, могут быть вызваны самыми разными причинами. С ними нужно быть осторожным. Например, если парень лайкает окружающих мало, но, при этом, какой-то девушке ставит лайков довольно много, наверное, в этом что-то есть. Либо симпатия, либо он просто написал бота, который ей эти лайки ставит. А может и то и то сразу. Я попробовал найти ответы на несколько подобных вопросов. В итоге получился довольно большой рисёрч. 

Абсолютно всё, что я сделал, я заливаю сюда. Все свои коды на python с подробными описаниями, все интерактивные визуализации, все тексты с описаниями. Я старался сделать всё это довольно подробным. Одним словом, добро пожаловать в гранд иканам рисёрч. Надеюсь, что он вам понравится. И для начала, я хотел бы поведать вам одну поучительную историю.

## 3 Древняя и поучительная история

В 1958 году была опубликована статья экономиста Филиппса, в которой он обнаружил достаточно чёткую отрицательную связь между инфляцией и безработицей в Англии за прошедшие 70 лет. Проверка этой работы на американских данных подтвердила наличие взаимосвязи. Так родилась знаменитая кривая Филиппса. Её стали интерпретировать как некую возможность выбора между высокой инфляцией и высокой безработицей. Нужно сократить безработицу? Давайте напечатаем денег и разгоним инфляцию!

Любому политику безработица кажется более значимой социальной проблемой, нежели инфляция, и он хочет её победить любыми доступными средствами. Самым популярным средством по борьбе с безработицей является агрессивная монетарная политика, которая состоит в расширении денежной массы. Именно это и было сделано президентом США Ричардом Никсоном в начале 1970-х гг. в ходе погони за низкой безработицей и высокой инфляцией.

К сожалению, план Никсона удался только наполовину, он добился высокой инфляции, но сбить безработицу не смог. В краткосрочном периоде она действительно упала. Через некоторое время вернулась назад. Кривая Филиппса в этот период времени исчезла из данных. Никой взаимосвязи между безработицей и инфляцией больше не прослеживалось.

В течение 70-х все Западные страны поэкспериментировали с кривой Филиппса. В конечном итоге, 1980-е годы были потрачены на обуздание разогнанной инфляции. Победа над инфляцией обошлась довольно дорого. Например, Пол Волкер, возглавлявший в этот период времени ФРС США, взвинтил ставку процента до 19%. Конечно же, это отправило экономику в рецессию, однако инфляционные ожидания были сбиты.

**Какова мораль этой истории?** Пацаны, начав бороться с безработицей, печатая деньги, чуток попутали причину и следствие. На самом деле никакой чёткой отрицательной зависимости между инфляцией и безработицей в природе не существует. Её наличие в данных является следствием сложного взаимодействия между экономическими агентами. Попытавшись воздействовать на следствие, а не на причину, политики налажали.

В ходе гранд иканам рисёрча мы тоже найдём некоторые закономерности. Из этих закономерностей можно будет сделать либо правильные, либо не очень правильные выводы. Давайте посмотрим на то как это работает.

###  3.1 Пример неправильного вывода: 

> Что что что? Вероятность того, что меня отчислят очень высока? Наличие подписок на мусорные паблики увеличивает эту вероятность? ЭЙ, АВТОР, ПЕРЕДЕЛЫВАЙ ПРОГНОЗЫ! Я ОТПИСАЛСЯ ОТ МДК! 

Нет. Это работает не так. Вы не знаете о том, что по вам делают рисёрч. Вы ведёте в сети какую-то активность. В ходе этой активности вы оставляете некоторый след о своих умственных способностях и особенностях характера. Наличие у вас в подписках МДК — это последствие этой активности. То есть это всего лишь инструмент, который сигнализирует, что у вас __уровень IQ ниже,__ чем у других людей. В ходе рисёрча это становится ясно и модель делает прогноз. 

Как только вы узнали о том, что наличие подписки на МДК выдаёт вас, вы, конечно же решите отписаться от него. Тем не менее, это не увеличит ваши шансы сдать сессию. Просто-напросто вы лишите инструмент, через который модель узнала о вас что-то, своей силы. Не путайте причину и слествие. __Ваш IQ низкий => Вы подписались на мусорный паблик => Модель заметила, что у вас низкий IQ и спрогнозировала высокую вероятность отчисления.__ Но ни в коем случа не наоборот. Отписавшись, вы засорите канал, по которому идёт информация, а не станете умнее. 

### 3.2 Пример правильного вывода: 

> Что что что? Вероятность того, что меня отчислят очень высока? Наличие МДК в подписках увеличивает эту вероятность? Плевать я хотел на МДК! Пойду ботать! Ботать это круто! 

Да. Это работает именно так. Вы воздействуете не на следствие того, что у вас низкий IQ, а на причину этого. Именно так и нужно делать. 

### 3.3 Другой пример неправильного вывода: 

> Такс, такс, такс, рисёрч, ага. У меня довольно низкие шансы быть отчисленым. Ну и нафиг всё это ботанье, я пошёл в кино. 

Нет. Это работает не так. То, что модель прогнозирует вам высокий шанс выживания, не означает, что можно ничего не делать. Вы правильно вели себя до этого и усердно трудились. Через разные инструменты, собранные для того, чтобы отследить это (вроде тех же подписок), модель поняла, что на этой сессии вы также будете стараться и приклаыдвать услия. Работает следующая цепочка. __Он был усерден => он оставил вк след, какой обычно оставляют усердные люди => судя по всему, он будет усердным и в этот раз => низкая вероятность отчисления__  И ни в коем случае не наоборот! 

### 3.4 Другой пример правильного вывода: 

> Такс, такс, такс, рисёрч, ага. У меня довольно низкие шансы быть отчисленым. Это круто. Можно пойти и как следует поботать. 

Да. Это работает именно так. Здравые мысли здравого человека. 


### 3.5  Ещё один неверный вывод: 

> Ага. Граф друзей. Ага, судя по результатам алгоритма PageRank я не очень авторитетный чувак на своём курсе. Этот алгоритм смотрит на то, кто у меня в друзьях. Добавлю ка я к себе в друзья пару авторитетных пацанов и получу [Признание.](http://bestgames4all.ru/uploads/images/g/t/a/gta_sa_missija_vipolnena.jpg)

Нет, это работает не так. __Вы социально значимый человек  => вы общаетесь с другими социально значимыми людьми => вы, возможно, добавляете их в друзья => у вас высокий PageRank.__  Социальная значимость человека - ненаблюдаемая переменная. Чтобы как-то учесть её при исследовании, приходится искать какие-то другие переменные, которые с ней коррелируют. Такие переменные называются инструментами. PageRank, построенный по тому какие именно у человека друзья, коррелирует с его социальной значимостью. 

Ещё раз, ещё раз. Он ни в коем случае не является эквивалентом социальной значимости, он лишь несёт о ней некоторую информацию. Искуственно завысив какие-то показатели, по которым можно судить о вашей важности, вы не увеличите свою важность. 

### 3.6  Ещё один неверный вывод: 

> Ага. Граф друзей. Ага, у меня высокий PageRank. Я очень крут. 

Нет, это работает не так. То, что вы попали в топ по одному какому-то алгоритму, не означает, что вы круты.

### 3.7  Ещё один верный вывод: 

> Ага. Граф друзей. Ага, у меня высокий Pagerank. Это интересно. Довольно мило, что данный алгоритм сигнализирует о том, что у меня высокий авторитет. Это не повод зазнаваться. Возможно, у меня просто очень много друзей. Особенно тех, с которыми я не общаюсь и в данной ситуации эта инструментальная переменная сработала не очень хорошо. ### 3.8 Ещё один неверный вывод: 

> Юля очень классная девушка и нравится мне. Вася поставил ей за прошедшие 2 года огромное количество лайков. Пора набить ему морду. Он хочет отбить её у меня. 

Нет, это работает не так. Количество лайков это интересная метрика. Да, она коррелирует с симпатиями, которые являются ненаблюдаемой переменной, однако бывают люди, которые лайкают всё подряд. Тем не менее, это не мешает им получать интересные результаты и лучше понимать то, как именно великий процесс порождения данных задумал окружающий нас мир. Я вовсе не спорю с тем, что 400 наблюдений это мало и о [законе больших чисел](https://ru.wikipedia.org/wiki/Закон_больших_чисел) говорить сложно.

Маленькое количество данных очень сильно связывает нам руки и ограничивает нас классом довольно простых моделей. Мы не можем строить модные случайные леса или нейросетки, однако это вовсе не повод сидеть сложа руки. Четыре сотни — это на порядок больше, чем 20. Если в данных хоть что-то есть, мы худо-бедно найдём это и используем. Да и к тому же как только кого-нибудь из читателей отчислят, появятся новые данные, и модели можно будет обучить на свеженькой информации.

## 4 Карта гранд рисёрча 

Надеюсь, что вы внимательно прочли пятый пункт и вникли в то, что нельзя путать причину и следствие, а также поняли что такое ненаблюдаемая переменная и что такое инструмент. Если да, то вы готовы двинуться в путь. Однако ни один путь не возможен без карты. Все файлы в иканам гранд рисёрч делятся на 3 типа. 

1. Код: это юпитерская тетрадка с кодом на питоне и подробными комментариями к нему. 
2. Почиташка: это юпитерская тетрадка с описанием результатов исследования.
3. Визуализашка: это интерактивная визуализация.

Иногда будут встречать гибриды. Например, Код-Почиташка. В таких файлах есть и питоновский код и подробное описание результатов исследования. Ко всем данным прилагается короткая справка по файлам в файликах `readme`. Если вы не собираетесь разбираться в коде, просто игнорируйте его, читайте текст и смотрите на картинки. Приятного чтения. 

###  Часть нулевая.

Эта часть находится здесь для тех, кто собирается досконально разбираться в питоновском коде. В первом блокноте рассказывается как в принципе происходит парсинг сайтов в python. В блокноте скачиваются мемы. После того, как сайт начинает нас блокировать, подключается Tor. Во втором блокноте подробно рассказывается о том как можно скачивать данные из вконтакте. В качестве иллюстрации приводится три примера: скачка паблика с мудрыми цитатами, скачка друзей для графа друзей, скачка лайков, которые друзья ставят друг-другу. 

* **Код:** [Туториал. Как раздобыть немножко данных о мемах.](https://habrahabr.ru/company/ods/blog/346632/) Наша с Димой статья на Хабре про сбор данных и обход блокировок в целом. Если вы уже немного знаете питон и хотите начать собирать данные, она неплоха для старта. 
* **Код:** [Туториал. Как скачать данные из контакта с помощью python.](http://nbviewer.jupyter.org/github/FUlyankin/ekanam_grand_research/blob/master/0.%20vk_parser_tutorial.ipynb) Продолжить разбираться в коде стоит туториалом по скачке любых данных из контакта.

> Внимание! Визуализация с тем кто кого лайкает сделана в R-shiny. На одном аккаунте может находиться не более 5 приложений. Приложение суммарно можно эксплуатировать не более 25 часов в сутки. Если при переходе по ссылке, вы видите ошибку, визуализация упала из-за очень большого числа людей, которые хотят с ней поиграться. Если вы хотите узнать кто кого лайкает вне очереди, имеет смысл скачать либо архив с R, либо [специальный архив с данными по лайкам и питоновской тетрадкой.](https://yadi.sk/d/cGSR7a003RUnnn) Rод для лайков довольно простой.
