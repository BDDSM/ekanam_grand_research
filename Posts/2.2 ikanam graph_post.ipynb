## 1.2 Центр графа иканама 

Познакомимся с графом иканама поближе.  Для этого посмотрим на основную статистику по нему. Заодно повторим дискретку. В нашем графе оказалось 636 вершин. Между ними проложено 11917 рёбер. Это означает, что между 636 людьми существует 11917 дружеских связей. Логично, что всего может существовать $С_{636}^2 = 201930$ дружеских связей. Выходит, что эконом дружит не особо плотно.

Начинаем повторять определения!  **Диаметром графа** называется максимальное расстояние между двумя его его вершинами. В нашем случае он равен $7$. Это означает, что для иканама работает теория 7 рукопожатий. Все люди знакомы друг с другом через 7 человек. Именно так Facebook на основе данных из своей социальной сети [проверил теорию 6 рукопожатий.](https://meduza.io/feature/2016/02/04/v-druzyah-u-druga-druga-druga)

**Эксцентриситет вершины графа** —  это расстояние до максимально удаленной от нее вершины. **Радиус графа** — это наименьший из эксцентриситетов всех вершин. В нашем случае радиус графа равен $4$. 

Первым способом нащупать с помощью графа связей самых важных людей, является концепция центра графа. **Центр графа** — это любая вершина, такая, что расстояние от нее до наиболее отдаленной вершины минимально. Центром графа может быть одна вершина или несколько вершин. Иначе говоря, центр графа — вершина, эксцентриситет которой равен радиусу. В случае иканама центровыми оказываются шесть человек. 

<img align="center" src="../images/2. loises_and_graph/gr_center.png" width="350">

**Коэффициентом кластеризации** называют степень вероятности того, то два разных пользователя, связанные с конкретным, тоже связаны. Этот показатель характеризует то, насколько плотно в графе расположены рёбра. В нашем случае он равен $0.55$. Это означает, что если мы возьмём двух друзей какого-нибудь человека, вероятность, что они дружат равна $0.55$.


## 1.3 Pagerank 

Перейдём к более интересным концепциям и попробуем проанализировать авторитетность с помощью алгоритма, который носит гордое название [PageRank.](https://ru.wikipedia.org/wiki/PageRank) Именно этот алгоритм Google использовал в донейросетевую эпоху при ранжировании сайтов в поисковой выдаче. Сегодня этот алгоритм является одной из состовляющих гугловской поисковой махины. 

Все страницы в интернете связаны между собой гиперссылками. Значит можно представить его в виде ориентированного графа. Если с одной страницы можно перейти на другую, то рисуем ребро. Логично, что чем больше гиперссылок указывает на страницу, тем она авторитетнее. Давайте откроем браузер и заставим его случайно блуждать по интернету. Он открывает какую-то страничку, на ней есть несколько гиперссылок. Он случайно выбирает одну из ссылок и переходит по ней, на новой странице он повторяет свой выбор и так далее. Представим себе, что наш браузер блуждает по интернету уже довольно давно. За свою жизнь он повидал каждую страницу десятки, а то и сотни раз. На какие-то страницы он попадал чаще, на какие-то реже. В итоге он провёл на каждой странице какую-то определённую долю своего времени. Чем выше эта доля, тем важнее страница. 

Сделаем аналогичную штуку с графом иканама. Запустим в нём случайное блуждание по страницам и посмотрим какую долю времени на чьей странице он проведёт. 


<img align="center" src="../images/2. loises_and_graph/pagerank.png" width="1200">


Обратите внимание, что доля времени, которую блуждание проводит на странице человека не превышает сотой. В сумме все доли дают единицу. Логично, что при росте количества друзей, доля времени, которую мы будем проводить на странице человека, будет возрастать. Однако, если мы добавляем в друзья более социально впутанного человека, то наш PageRank возрастает сильнее. 

Например, если некоторый первак добавит к себе в друзья Никиту Арбатского, его PageRank увеличится сильнее, чем если он добавит к себе в друзья Никиту Фокина. Добавив в друзья человека, он добавит себя во все маршруты, которые проходят через его друзей. Это позволит увеличить долю пребывания на своей странице пропорционально PageRank добавленного человека.

Авторитет человека внутри социальной группы - ненаблюдаемая переменная. Для её описания необходимо придумывать какие-то иснтрументы. Pagerank показатель является одним из таких инструментов. С тем насколько он хорошо работает, можно спорить. Если человек авторитетен внутри своего сообщества, он должен быть как следует в него вписан. Именно эту вписанность пытается уловить алгоритм. К сожалению, многие люди добавляются в вк друг к другу в друзья просто так. Это зашумляет результаты. Мы попробуем очистить наш граф от таких случайных связей на основе лайков, которые люди ставят друг-другу. Тем не менее, такой алгоритм также будет не очень хорош. Истиные авторитеты ставят лайки очень редко. Если у вас есть свои идеи, связанные с тем как можно измерить авторитет человека в рамках социальной группы, я буду рад с вами их обсудить.

# 2. Графы лайков 

Добавить к себе в друзья можно кого угодно. Дружат ли люди на самом деле — большой вопрос. Если люди оставляют друг-другу хоть какое-то количество лайков, видимо, между ними есть дружба. Попробуем построить на основе матрицы лайков, которую мы получили в предыдущем скрипте, граф лайков. Обратите внимание на то, что  этот граф окажется ориентированным. Лайки ставятся в одностороннем порядке.  Весами рёбер в нашей матрице будет количество лайков, которое один юзер поставил другому. Если прочитав это, вы подумали: "Зачем ставить веса?!", то вы молодец и многое поняли, прочитав предыдущую часть рисёрча. Учёт весов, действительно приведет к проблемам. Сам по себе граф можно нарисовать вот так: 


<img align="center" src="../images/2. loises_and_graph/dir_gr.png" width="400">


Как уже говорилось выше, симпатии между людьми и возникающие из-за симпатий социальные связи — ненаблюдаемые переменные. Лайки и добавления друг-друга в друзья коррелируют с этой скрытой переменной. Из-за этого они могут использоваться как инструменты для изучения сообществ людей. Попробуем запустить PageRank на нашем новомодном ориентированном графе. Когда мы находимся в некоторой вершине, ребро для дальнейшего перехода будет выбираться  с вероятностью, пропорциональной весу это ребра. 

<img align="center" src="../images/2. loises_and_graph/likes_pagerank_1.png" width="1100">

По некоторой причине, связанной с самым качетственным пивом в Москве, возникают сомнения в качестве работы алгоритма. Социально важными сразу же оказываются люди, которых очень часто лайкают. В том числе те, кто часто лайкает сам себя. Это не очень круто. Можно попробовать по аналогии сделать PageRank по лайкпостам.

В топ, в таком случае начнут выходить люди с не очень большим числом постов. Выходит, что работа с лайками в виде графа сопряжена точно с такими же проблемами, как и составление рейтингов по матрице. Это неудивительно, так как по сути PageRank ищет похожие закономерности.

Какая разница сколько лайков ставит человек? Самое главное, что лайк был поставлен. Это означает, что страница привлекла человека. При этом она привлекла его в одностороннем порядке. То, в каком объёме она его привлекла, на самом деле не очень важно. Попробуем подойти к исследованию немного иначе. Возьмём матрицу из лайков и заменим в ней количество лайков, которое ставит человек на единицу, если хотябы один лайк был поставлен и на ноль, если ни одного лайка поставлено не было.


<img align="center" src="../images/2. loises_and_graph/likes_pr_2.png" width="1200">

Теперь давайте задумаемся вот о чём: одного лайка мало для возникновения симпатии. При этом сто лайков - слишком много. Возникает вопрос: а сколько достаточно? На этот вопрос нет однозначного ответа.

Подняв порог очень сильно, мы наткнёмся на то, что итоговый граф станет очень сильно разряжен. В итоге останутся только залайканные люди и мы вновь наткнёмся на ту же самую проблему, от которой пытаемся уйти. При этом, для людей, которые делают очень мало постов и один лайк является критичным. Если вы попробуете перебирать пороги, вы неожиданно для себя обнаружите, что алгоритм работает неустойчиво и вытаскивает в топ людей полурандомно. Идея заменить а матрице дружеские связи на лайки потерпела крах.

Тем не менее, это не мешает вам попытаться развить её в каком-нибудь интересном направлении. Например, можно попробовать построить граф на основе матрицы из f-мер и также постараться подобрать порог. Возможно, что результат окажется более устойчивым.
