## 1.2 Центр графа иканама 

Познакомимся с графом иканама поближе.  Для этого посмотрим на основную статистику по нему. Заодно повторим к экзамену по дискретке, если мы с первого курса, основные понятия. В нашем графе оказалось 636 вершин. Между ними проложено 11917 рёбер. Это означает, что между 636 людьми существует 11917 дружеских связей. Логично, что всего может существовать $С_{636}^2 = 201930$ дружеских связей. Выходит, что эконом дружит не особо плотно.

Начинаем повторять определения!  **Диаметром графа** называется максимальное расстояние между двумя его его вершинами. В нашем случае он равен $7$. Это означает, что для иканама работает теория 7 рукопожатий. Все люди знакомы друг с другом через 7 человек. Именно так Facebook на основе данных из своей социальной сети [проверил теорию 6 рукопожатий.](https://meduza.io/feature/2016/02/04/v-druzyah-u-druga-druga-druga)

**Эксцентриситет вершины графа** —  это расстояние до максимально удаленной от нее вершины. **Радиус графа** — это наименьший из эксцентриситетов всех вершин. В нашем случае радиус графа равен $4$. 

Первым способом нащупать с помощью графа связей самых важных людей, является концепция центра графа. **Центр графа** — это любая вершина, такая, что расстояние от нее до наиболее отдаленной вершины минимально. Центром графа может быть одна вершина или несколько вершин. Иначе говоря, центр графа — вершина, эксцентриситет которой равен радиусу. В случае иканама центровыми оказываются шесть человек. 

<img align="center" src="../images/2. loises_and_graph/gr_center.png" width="350">

**Коэффициентом кластеризации** называют степень вероятности того, то два разных пользователя, связанные с конкретным, тоже связаны. Этот показатель характеризует то, насколько плотно в графе расположены рёбра. В нашем случае он равен $0.55$. Это означает, что если мы возьмём двух друзей какого-нибудь человека, вероятность, что они дружат равна $0.55$.


## 1.3 Pagerank 

Перейдём к более интересным концепциям и попробуем проанализировать авторитетность с помощью алгоритма, который носит гордое название [PageRank.](https://ru.wikipedia.org/wiki/PageRank) Именно этот алгоритм Google использовал в донейросетевую эпоху при ранжировании сайтов в поисковой выдаче. Сегодня этот алгоритм является одной состовляющих гугловской поисковой махины. 

Все страницы в интернете связаны между собой гиперссылками. Значит можно представить его в виде ориентированного графа. Если с одной страницы можно перейти на другую, то рисуем ребро. Логично, что чем больше гиперссылок указывает на страницу, тем она авторитетнее. Давайте откроем браузер и заставим его случайно блуждать по интернету. Он открывает какую-то страничку, на ней есть несколько гиперссылок. Он случайно выбирает одну из ссылок и переходит по ней, на новой странице он повторяет свой выбор и так далее. Представим себе, что наш браузер блуждает по интернету уже довольно давно. За свою жизнь он повидал каждую страницу десятки, а то и сотни раз. На какие-то страницы он попадал чаще, на какие-то реже. В итоге, он провёл на каждой странице какую-то определённую долю своего времени. Чем выше эта доля, тем важнее страница. 

Сделаем аналогичную штуку с графом иканама. Запустим в нём случайное блуждание по страницам и посмотрим какую долю времени на чьей странице он проведёт. 


<img align="center" src="../images/2. loises_and_graph/pagerank.png" width="1200">


Обратите внимание, что доля времени, которую блуждание проводит на странице человека не превышает сотой. В сумме все доли дают единицу. Логично, что при росте количества друзей, доля времени, которую мы будем проводить на странице человека, будет возрастать. Однако, если мы добавляем в друзья более социально впутанного человека, то наш PageRank возрастает сильнее. 

Например, если некоторый первак добавит к себе в друзья Никиту Арбатского, его PageRank увеличится сильнее, чем если он добавит к себе в друзья Никиту Фокина. Добавив в друзья человека, он добавит себя во все маршруты, которые проходят через его друзей. Это позволит увеличить долю пребывания на своей странице пропорционально PageRank добавленного человека.

Авторитет человека внутри социальной группы - ненаблюдаемая переменная. Для её описания необходимо придумывать какие-то иснтрументы. Pagerank показатель является одним из таких инструментов. С тем насколько он хорошо работает, можно спорить. Если человек авторитетен внутри своего сообщества, он должен быть как следует в него вписан. Именно эту вписанность пытается уловить алгоритм. К сожалению, многие люди добавляются в вк друг к другу в друзья просто так. Это зашумляет результаты. Мы попробуем очистить наш граф от таких случайных связей на основе лайков, которые люди ставят друг-другу. Тем не менее, такой алгоритм также будет не очень хорош. Истиные авторитеты ставят лайки очень редко. Если у вас есть свои идеи, связанные с тем как можно измерить авторитет человека в рамках социальной группы, я буду рад с вами их обсудить.

## 3. Красивые визуализации графов

Пакет `NetworkX` даёт довольно тошнотворные визуализации. Конечно же можно как следует поднапрячься, засесть на пару суток и родить граф-конфетку внутренними средствами пакета. Однако мы пойдём другой дорогой.

Если как следует погуглить, то можно найти [вот такое вот офигенное приложение](http://www.yasiv.com/vk) для контакта, которое строит граф друзей. Только дай ему доступ и на выход ты получишь вот такую офигительную интерактивную, быстро работающую штуку. 

<img align="center" src="../images/2. loises_and_graph/d3_fr_gr.png" width="900"> 

<br>

Можно добавлять людей, рскрашивать рёбра в другие цвета и всё такое. Можно даже выкинуть какого-нибудь человека и посмотреть на то как граф разваливается на части и освободившиеся вершины разлетаются в разные стороны. Эта визуализация богоподобна, а её автор гений.  Мой горький опыт говорит мне о том, что каждая хорошая визуализация это великое страдание. Чтобы построить такую прекрасную визуализацию, автор страдал довольно долго. Он писал на java script для визуализации [свою библиотеку.](https://github.com/anvaka/VivaGraphJS) На хабре можно найти [его статью](https://habrahabr.ru/post/144758/) об этом. 

Мы попробуем родить нечто похожее. К моему превиликому сожалению, я не могу в javascript, поэтому мы будем рожать нечто похожее по уже готовым шаблонам. Из статьи выше можно узать, что основным тормозом подобных визуализаций является браузерный код отрисовки элементов. Автору удалось перетащить это бутылочное горлышко в свой алгоритм укладки. Благодаря этому его визуализация работает быстро. Это плохая новость для нас. Наш шаблон будет хорош только для маленьких графов. Большие будут сильно тормозить.

В сети выложено довольно много [шаблонов с красивым интерактивом.](https://github.com/d3/d3/wiki/Gallery) Люди, которые выкладывают эти визуализации в сеть - прекрасны. Они делают нас на шаг ближе к дзену, ровно как и команда `import this`. 

Нас будет интересовать [шаблон графа.](https://bl.ocks.org/mbostock/4062045) В конечном счёте, мы сможем открывать его в отдельной вкладке и дёргать за вершины. Возьмём шаблон и разберёмся в том как его правильно использовать. На страничке с описанием графа есть две вставки. Одна из них называется index.html 

<img align="center" src="../images/2. loises_and_graph/d3_index.png" width="550"> 

В этой вставке задаётся внешний вид графа и его моторика. Если вы знаете javascript или просто (как я) обладаете хорошей интуицией и можете в метод тыка и гугл, вы можете видоизменить дизайн графа.  Вторая вставка называется miserables.json.

<img align="center" src="../images/2. loises_and_graph/miserables.png"  width="550"> 

В этой вставке находится содержимое графа. В виде словарика задаётся какие вершины с какими соединяются и как подписываются. Эти два куска кода можно скопировать к себе на рабочий стол в два разных текстовых файла. 

<img align="center" src="../images/2. loises_and_graph/d3_desktop.png" width="300"> 

Теперь, если открыть файл index.html в браузере firefox, вашему взору предстанет та самая визуализация с сайта. 

<img align="center" src="../images/2. loises_and_graph/d3_firefox.png" width="700"> 

Наша задача заключается в следующем: мы должны сгенерировать словарик со своими друзьями и связями между ними, а после закинуть его в файл miserables.json. Таким образом, интерактивная визуализация будет готова и мы даже сможем поделиться ей со своими друзьями.

Итак, сгенерируем словарик следущего вида: 

```
{ 
"nodes": [ 
        {"id":"Aлександр Конев", "group":1}, 
        {"id":"Александр Тишин", "group":1}, 
        {"id":"Давыдкина Кристина","group":1} 
        ], 
"links":[ 
        {"source":"Aлександр Конев", "target":"Александр Тишин", "value":1}, 
        {"source":"Aлександр Конев", "target":"Давыдкина Кристина", "value":8} 
        ] 
}
```

Первый список - вершины. Цифра в `group` отвечает за цвет вершины. Второй список - рёбра. `source` и `target` означают вершины, которые нужно соединить ребром. Цифра в `value`, судя по всему, отвечает за толщину ребра. В принципе, построить такой текстовый файлик можно даже вручную...

Как только мы сделаем это и откроем наш новоиспечённый граф иканама в браузере (желательно firefox), мы увидим желаемое. Жёстко подвисающее, долго строющееся, но желаемое. 

<img align="center" src="../images/2. loises_and_graph/d3_ekanam_1.png" width="700"> 

Как я и обещал, работать очень медленно. Этот граф лежит в репозитории в папке ekanam_gr_1. При желании его можно скачать по ссылке, которая будет ниже и упороться. 

Попробуем спуститься на уровень ниже и сделаем интерактивным более маленький граф. Ко всем вершинам не забудем сделать подписи. Для этого подберём [новый шаблон](https://bl.ocks.org/heybignick/3faf257bbbbc7743bb72310d03b86ee8) Забираем с сайта новый `index.html` и генерируем для него точно такой же как и раньше `miserables.json`. Внутри index.html методом тыка догадаемся где надо подписать хорошие цифры, чтобы граф стал более разряженым.

Вершины можно дёргать и вертеть, а картинку не стыдно вставить в свой НИР. Тем не менее, есть куда расти. Следующим шагом я хотел заменить все вершины на фотки. Делать это я собирался на основе html из вот этой [статейки с хабра.](https://habrahabr.ru/post/221251/)  Cоответствующий html шаблон лежит в [репозитории, в папке web.](https://github.com/stleon/vk_friends) Однако мне не удалось заставить этот шаблон работать в течение часа с лишним и я на него забил. Следующим шагом может стать раскраска веток по аналогии с тем как это сделано в визуализации друзей выше, но для этого надо мочь в javascript.
