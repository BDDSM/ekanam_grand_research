# 1. Про ошибки, которые мы допускаем 

Скачать контент всех пабликов, а после скачать все лайки, которые стоят под каждым постом, дело нетрудное, но времязатратное. Около 6 часов уйдёт на то, чтобы скачать из 20000 пабликов по 100 постов. После надо будет к каждому из 200000 постов обратиться по несколько раз (в зависимости от числа лайков), чтобы собрать списки из id, оставивших своё мнение под постом. На это, при использовании метода `execute`, который, грубо говоря, позволяет делать запросы в 25 раз быстрее, уйдёт около 18 часов.

Код для парсинга получится совсем небольшим. Когда он будет дописан, вы включите свой старый ноут, поставите на нём код (новый ноут вы возьмёте с собой на пары/работу/тусу), увидите вот такую, радующую глаз картину и оставите расчёты на произвол судьбы. 

<img align="center" src="../images/3. publics/likes.png" width="1200">






Вроде бы код работает нормально, все проверки показали, что он, действительно, скачивает разные лайки с разных групп, под разными постами. Вы однажды уже ставили расчёты на 6 часов, а потом неожиданно обнаружили, что код скачал $20000$ раз одно и то же. Именно тогда вы прокричали своё **первое даташрушерское блять.** Сегодня вы такого не допустите. Код проверен много раз и работает корректно. Интересно, что же мы увидим через $18$ часов...

Мы увидим зависший экран. Неожиданно обнаружится, что вашему старому ноуту уже 6 лет, на нём очень мало оперативки и с новым маком, на котором вы готовили этот код, он соперничать в плане производительности не может. Всё зависло на $20\%$. НУ КОНЕЧНО! Надо было внутри кода прописать, чтобы данные сохранялись в новый файлик каждые $10\%$. Тогда бы это зависание не было бы таким обидным, но вы это не сделали. Вы рассчитывали, что эти $18$ часов будут отработаны кодом также чётко, как были отработаны предыдущие $6$. Вместо этого всего вы уже орёте во всё горло: **БЛЯТЬ!** 

Не беда, завтра же выходной. Вы весь день планировали протусить дома. Можно запустить расчёты на маке на всю ночь и весь день. Перезапускаем код на маке, убеждаемся что всё окей и, когда счётчик доходит до $5\%$, отправляемся спать. Вот смеху то будет утром, когда вы проснётесь и неожиданно выясните, что забыли обновить ключ доступа к API. А он, между прочим, выписывается на $24$ часа. Особенно забавно то, что он истёк ровно через $10$ минут после того как вы уснули. Весь экран заполонит полчище сообщений в стиле `Проблемы с 228 батчем`. Что будет дальше? Вы, как настоящий дата сайнтист, проорёте во всё горло то самое сокровенное и неистовое **БЛЯТЬ,** обновите ключ доступа и перезапустите код. 

Третья попытка увенчается успехом. Всё скачалось. Гигабайты из лайков лежат на жёстком диске вашего компьютера. Вместо одних суток, которые вы планировали потратить на скачку, вы потратили трое. Это только скачка. Данные надо будет ещё обработать, после надо будет обучить несколько моделей и получить какие-то результаты. Конечно же вы попытаетесь предотвратить все эксцессы и сначала попробуете запускать код на маленьких подвыборках, пока не окажетесь уверены в своём коде на все $99%$. Однако ваша предусмотрительность переодически будет давать сбои. Более того, все те данные, которые вы скачали, можно умудриться каким-нибудь неведомым образом повредить. По крайней мере, у меня получилось. После этого придётся всё перекачивать заново и делать резервную копию данных, на всяий случай. 

Тернист и сложен путь человека, решившего работать с данными. Если вы приняли такое решение, то готовьтесь к довольно широкому спектру ошибок, которые вы неожиданно для себя допустите. Потом можно будет рассказать о своей неимоверной тупости друзьям и услышать в ответ их истории. Также можно будет всем вместе купить измерители уровня шума, развесить их в своих комнатах и сравнивать чьё **даташрушерское блять** оказалось громче. Данные приходят и уходят, модели обучаются и запускаются в прод, а наше блять, всегда остаётся с нами.
