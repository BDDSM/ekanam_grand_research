## О том, что остаётся в интернете

Одним прекрасным утром я проснулся у себя дома и пошёл на кухню, чтобы заварить себе чай и взять пару пироженок. Попивая чаёк я планировал просматривать мемы и читать вконтакте иканамовский чатик. Утро было спокойным и безмятежным. Ничего не предвещало беды. Когда я открыл чатик, я увидел как пацаны над чем-то угарают. "Должно быть новый мемас",- подумал я. Это был не мемас. Промотав чатик чуток повыше, я увидел нечто из прошлого. И, конечно же, орнул. В чатике лежал следующий скриншот:

В пиковый период своей зелёной деятельности, я нашёл на каком-то левом сайте проект с капсулами времени и запилил послание потомкам, которые без моей мудрости явно бы не выжили. По сравнению с тем, что ты ежедневно отдаёщь в распоряжение большим корпорациям вроде Мэйла или Гула, то, что я забрал у тебя, довольно безобидная штука. ## О Куполе, под которым мы сидим 

Чтобы узнать в красках о том, какой инфомрационный след ты за собой оставляешь, можно посмотреть [видос одного блогера на ютубе.](https://www.youtube.com/watch?time_continue=1&v=swnnpqu7a6s) Рассказано красиво, в эмоцях и немного натянуто.

Откровенно говоря, любому человеку, который работает с даннымми, всё это покажется очевидным. Более того, с частью тезисов даже можно поспорить. Тем не менее, общая картина действительно выглядит примерно так. Машинное обучение, в купе с твоим информационным следом, позволяет любой крупной компании знать о тебе довольно многие вещи.

Например, машинное обучение может помочь [магазину узнать о вашей беременности раньше, чем об этом узнает ваш отец](http://ru.datasides.com/магазин-узнал-о-беременности-девушки/) и даже ненароком сдать вас ему. Оно может помочь вам найти свой собственный [закон пива и подгузников](http://tmstudent.ru/article/keis-pampersy-i-pivo) и повысить прибыль. Более того, оно может помочь вам грамотно сагетировать избирателей и [одержать на выборах победу.](https://theins.ru/politika/38490) Вокруг последних выборов идут довольно жаркие споры. Однако, прочитав книгу [Верховный алгоритм](https://www.ozon.ru/context/detail/id/136780904/), можно узнать, что судьбу США в 2012 году определило именно машинное обучение. Обама оказался первопроходцем в использовании больших данных в политике.

Иногда сбор данных происходит очень неожиданным образом. Например, сегодня социальной сетью вконтакте владеет Mail.Ru Group. Не так давно он [начал эксперементировать с психотипами людей.](https://www.youtube.com/watch?time_continue=1&v=swnnpqu7a6s) В соц сети  было запущено приложение-тест, пройдя который, пользователь мог узнать немного больше о своей психологии. Проходя этот тест, он невольно помогал мэйлу делать разметку выборки. Собрав достаточное количество данных, mail смог обучить модель, которая по соодержимому странчки человека понимает каким психотипом он обладает.

Сегодня данные о психотипах (и другие данные из соц сеток) обученные алгоритмы используют для оценки производительнсти потенциальных сотрудников. Вполне возможно, что ваше резюме будет отобрано или не отобрано их кипы других именно благодяря такому алгоритму. 

Только что приняли решение почистить свой профиль? [Поздно.](https://socialdatahub.com/ru/sbor_i_strukturirovanie_dannyih) Представляете себе сколько таких контор собирающих данные, ведут свою деятельность? Более того, в последние годы наличка уходит на задний план. Мы оплачиваем практически все транзакции банковскими картами. Это приводит к тому, что банкам доступна детализированная информация о том где мы и что покупаем. Как вы собираетесь удалить эти данные? 



## Зачем я это пишу

Если ты впечатлительный и никогда не слышал про анализ даных, моя небольшая сказка про анонимность, рассказанная выше, могла тебя напугать. На самом деле бояться нечего. Ну подумаешь, какой-нибудь банк купил информацию о том что ты гуглишь и использует её при решении выдавать тебе кредит или нет. Он использует её в совокупности с кучей других запросов. И единственное, что может произойти, это то, что аналитик поугарает над тем какое именно порно ты вчера искал, а потом дома, в режиме инкогнито, тоже попробует его посмотреть. Пока ты не нарушаешь закон, всем на тебя пофигу. А когда ты его нарушишь, тогда тебя, скорее всего, найдут. 

Я рассказываю всё это лишь для того, чтобы показать возможности. В ближайшее время [спрос на высококвалифицированную рабочую силу,](http://tass.ru/pmef-2016/article/3345977) к которой относятся и аналитики, будет расти. В данный момент в мире очень сильно не хватает людей, которые разбираются в алгоритмах машинного обучения. При этом спрос на специалистов со среденй квалификацией (бухгалтера, юристы и тп) будет падать. [Многие профессии исчезнут.](https://vk.com/ppilif?z=video-75410445_456239067%2F3c06fddeb99f78371a%2Fpl_post_-75410445_3892) Ключевым навыком в будущем будет уменее учиться, отсеивать самую важную информацию и не выпадать. Иначе ваша работа, также как работа 3000 юристов в сбере, рано или поздно [будет передана роботу,](https://www.rbc.ru/rbcfreenews/5877b2979a79478752358fb9) которого обучили дата сайнтесты. Не поймите меня неправильно. Юристы не исчезнут, они останутся. Но только самые высококвалифицированные. 

Будущее наступает. И вы можете не просто жить в нём, а творить его. Для этого осталось только выучить математику, научится программировать как минимум на двух языках, разобраться в кипе алгоритмов, а также разобраться какими бывают данные и понять как их можно добывать и хранить. Надеюсь, что моё небольшое исследование и все те ссылки, которые я привёл выше, промотивируют вас к этому.
