## 1 О том, что остаётся в интернете

**На всякий случай: синий текст - это гиперссылки. На них можно жать. Синий цвет как бы намекает, что вас перенесёт на другую страницу.** 

Одним прекрасным утром я проснулся у себя дома и пошёл на кухню, чтобы заварить себе чай и взять пару пироженок. Попивая чаёк, я планировал просматривать мемы и читать вконтакте иканамовский чатик. Утро было спокойным и безмятежным. Ничего не предвещало беды. Когда я открыл чатик, я увидел как пацаны над чем-то угарают. "Должно быть новый мемас",- подумал я. Это был не мемас. Промотав чатик чуток повыше, я увидел нечто из прошлого. И, конечно же, орнул. В чатике лежал следующий скриншот:

В 2009 году, когда я учился в 8 классе, я жёстко угарал по защите окружающей среды. У меня даже была группа вконтакте под названием "Мир" с диванной аналитикой по глобальному потеплению. В пиковый период своей зелёной деятельности я нашёл на каком-то левом сайте проект с капсулами времени и запилил послание потомкам, которые без моей мудрости явно бы не выжили. В итоге, спустя 8 лет, интернет ни о чём не забыл, и Сёма Осипов каким-то образом откопал этот сайт. **Всё, что попало в интернет, навсегда остаётся в интернете.** Каждый из нас оставляет за собой какой-то информационный след, и такие эпичные записи далеко не самая ценная его часть. 

Наш след помогает всем заинтересованным понять какие мы с вами на самом деле: насколько открытые, умные, социализированные, что любим покушать, что хотели бы купить и т.п., а после как-то использовать всю эту информацию. Например, посмотреть [насколько часто похожие на нас люди](https://www.rbc.ru/technology_and_media/06/10/2017/59d6b4309a79475f4c3e5c59) возвращают кредиты и решать на основе этого давать нам их или нет.

Даже если мы решим почистить за собой всё то, что наследили, это не особо поможет. В сети существует огромное количество различных [баз с данными,](https://socialdatahub.com/ru/sbor_i_strukturirovanie_dannyih) которые на ежедневной основе собирают всё то, что мы оставляем за собой. Более того, подобные базы есть у любой крупной компании. Всё это нужно, чтобы творить магию машинного обучения. 

Именно машинное обучение может помочь [магазину узнать о вашей беременности раньше, чем об этом узнает ваш отец](http://ru.datasides.com/магазин-узнал-о-беременности-девушки/) и даже ненароком сдать вас ему. Оно может помочь вам найти свой собственный [закон пива и подгузников](http://tmstudent.ru/article/keis-pampersy-i-pivo) и повысить прибыль. Более того, оно может помочь вам грамотно сагетировать избирателей и [одержать на выборах победу.](https://theins.ru/politika/38490) Вокруг последних выборов идут довольно жаркие споры. Однако, прочитав книгу [Верховный алгоритм](https://www.ozon.ru/context/detail/id/136780904/), можно узнать, что судьбу США в 2012 году определило именно машинное обучение. Обама оказался первопроходцем в использовании больших данных в политике. Кстати о политике. Кажется, [один парень](https://habrahabr.ru/company/fbk/blog/335778/) впечатлился всем этим. 

Иногда сбор данных о вас происходит очень неожиданным образом. Например, сегодня социальной сетью вконтакте владеет Mail.Ru Group. Не так давно он [начал эксперементировать с психотипами людей.](https://www.youtube.com/watch?time_continue=1&v=swnnpqu7a6s) В соц сети  было запущено приложение-тест, пройдя который, пользователь мог узнать немного больше о своей психологии. Попутно он рассказывал о своём психотипе мэйлу. Вся информация тщательно собиралась и в итоге, собрав достаточные объёмы данных, mail смог обучить модель, которая по соодержимому странчки человека прогнозирует его психотип. Сегодня эти данные используются для оценки работоспособности потенциальных сотрудников. Использует их не только мэйл. Вполне возможно, что ваше резюме будет отобрано или не отобрано из кипы других именно благодаря модели, обученой находить годных людей именно по этой выборке. 

Социальные сетки и поисковые запросы — не самое крутое, что можно найти. В последние годы наличные деньги уходят на задний план. Всё больше транзакций мы совершаем в одно касание с помощью банковских карт. Это приводит к тому, что банкам становится доступна детализированная информация о всех наших покупках. Узнать в красках о том, какие ещё информационные следы оставляют люди, можно в [видео одного блогера на ютубе.](https://www.youtube.com/watch?v=qBDbO_bAdFM&t=21s) Рассказано красиво, в эмоцях и немного натянуто. 

На самом деле бояться нечего. Ну подумаешь, какой-нибудь банк купил информацию о том что вы гуглите и использует её при принятии решения выдавать вам кредит или нет. Во-первых, база данных будет продана в анонимизированном виде. Какое наблюдение соответствует какой фамилии, будет знать только продавец. Во-вторых, поисковые запросы отдельного человека не очень сильно помогут при обучении модели. Данных нужно много. Да и единственное, что может произойти, это то, что аналитик поугарает над тем какое именно порно вы вчера искали, а потом,  дома, в режиме инкогнито, тоже попробует его посмотреть. Пока вы не нарушаете закон, всем на вас пофигу. А когда вы его нарушите, тогда вас, скорее всего, найдут. 

Если вы впервые в жизни про всё это слышите, должно стать немного жутковато. Тем не мене, я пишу всё это не с целью напугать, а с целью показать возможности и промотивировать ботать. В ближайшие годы [спрос на высококвалифицированную рабочую силу,](http://tass.ru/pmef-2016/article/3345977) к которой относятся и аналитики, будет расти. Спрос на специалистов со средней квалификацией (бухгалтера, юристы и тп) будет падать. [Многие профессии исчезнут.](https://vk.com/ppilif?z=video-75410445_456239067%2F3c06fddeb99f78371a%2Fpl_post_-75410445_3892) Если вы не сможете успевать бежать за временем, ваша работа, также как и работа 3000 юристов в сбере, рано или поздно [будет передана роботу,](https://www.rbc.ru/rbcfreenews/5877b2979a79478752358fb9) которого обучили дата-сайентисты. Не поймите меня неправильно. Юристы не исчезнут, они останутся. Но только самые высококвалифицированные.  

[Будущее наступает.](https://meduza.io/cards/ekonomisty-obsuzhdayut-chetvertuyu-promyshlennuyu-revolyutsiyu-chto-eto) И вы можете не просто жить в нём, а творить его. Сегодня в мире довольно сильно не хватает спецов по анализу данных. На то, чтобы всё это поместить в свою голову и как следует упорядочить, нужно убить уйму времени. Без должной мотивации ничего не выйдет. Поэтому главная цель моего рисёрча именно в этом: показать какие интересные штуки бывают на этом свете и замотивировать неистово ботать. 

## 2  О целях и задачах иканам гранд рисёрча 

В гранд эконом рисёрче мы будем исследовать данные об экономе. Перед тем как с головой в эти данные окунуться, стоит понять с какими целями мы туда окунаемся. Как мы уже выяснии, **главная цель:** замотивировать тебя ботать. Кроме этой главной цели есть ещё несколько.

#### Первая цель: 
Найти самых популярных и самых авторитетных людей на экономе. Попробовать найти взаимные симпатии. Мы попытаемся идентифицировать популярных и авторитетных людей с помощью дружеских связей и лайков. Также мы попробуем найти людей с самым интересным контентом. Для этого мы:

1. Скачаем все данные по дружеским связям и лайкам из контакта и приведём их к удобному виду.
2. Построим несколько лайк-топов и поймём чем эти лайк-топы плохи.
3. Придумаем свою собственную лайк-меру для поиска людей с интересным контентом.
4. Построим граф дружеских связей и граф лайков эконома.
5. Найдём самых авторитетных людей с помощью теории графов.

#### Вторая цель: 
Проанализировать паблики, на которые подписаны люди. Выделить тематики для пабликов, посмотреть какие тематики преобладают в профилях у каких курсов. Проанализировать популярность пабликов и выделенынх тематик. Для этого мы:

1. Скачаем все подписки ребят с эконома.
2. Скачаем по сто постов из каждого паблика.
3. Проведём кластеризацию пабликов по темам, которые в них поднимаются.
4. Нарисуем для каждого юзера свой собственный тематический профиль.

#### Третья цель:
Построить модель, которая могла бы спрогнозировать закончит ли поступивший студент первый курс. Делать это мы будем по данным из профиля вконтакте и по данным из приказов о зачислении. Для этого мы: 

1. Соберём по приказам о зачислении основные данные о людях, поступивших на эконом за последние пять лет.
2. Скачаем со страничек вконтакте всю доступную информацию. Как я уже говорил выше, заполненность нашего профиля и наши подписки - это всего лишь инструменты, которые коррелируют с нашими интересами и качествами. Инструменты помогут нам хотябы частично учесть в модели сложную структуру человеческого характера. Ниже мы поговорим про природу инструментов чуть более подробно. 
3. Переработаем все эти данные в красивые таблички.
4. Обучим классификатор, который попытается понять какие люди выживают на экономе, а какие нет. 
5. Построим прогнозы по отчислениям для новых первокурсников.

К сожалению, у нас в выборке окажется довольно мало наблюдений. Триста человек явно будет мало для того, чтобы модель смогла впитать в себя все грани сложного человеческого характера. Тем не менее, какие-то вещи она сможет понять. 

В ходе всей этой работы, мы поднимем на поверхность довольно большой пласт различных концепций. Иногда по тексту будут встречаться не только весёлые картинки, но и не менее весёлые формулы. Одним словом, добро пожаловать в гранд иканам рисёрч. Надеюсь, что он вам понравится. И для начала, для того, чтобы избежать возникновения в головах читателей неправильных выводов, я хотел бы поведать вам одну поучительную историю.

## 3 Древняя и поучительная история

В 1958 году была опубликована статья экономиста Филиппса, в которой он обнаружил достаточно чёткую отрицательную связь между инфляцией и безработицей в Англии за прошедшие 70 лет. Проверка этой работы на американских данных подтвердила наличие взаимосвязи. Так родилась знаменитая кривая Филиппса. Её стали интерпретировать как некую возможность выбора между высокой инфляцией и высокой безработицей. Нужно сократить безработицу? Давайте напечатаем денег и разгоним инфляцию!

Любому политику безработица кажется более значимой социальной проблемой, нежели инфляция, и он хочет её победить любыми доступными средствами. Самым популярным средством по борьбе с безработицей является агрессивная монетарная политика, которая состоит в расширении денежной массы. Именно это и было сделано президентом США Ричардом Никсоном в начале 1970-х гг. в ходе погони за низкой безработицей и высокой инфляцией.

К сожалению, план Никсона удался только наполовину, он добился высокой инфляции, но сбить безработицу не смог. В краткосрочном периоде она действительно упала, но через некоторое время вернулась назад. Кривая Филиппса в этот период времени исчезла из данных. Никой взаимосвязи между безработицей и инфляцией больше не прослеживалось.

В течение 70-х все Западные страны поэкспериментировали с кривой Филиппса. В конечном итоге, 1980-е годы были потрачены на обуздание разогнанной инфляции. Победа над инфляцией обошлась довольно дорого. Например, Пол Волкер, возглавлявший в этот период времени ФРС США, взвинтил процентную ставку до 19%. Такие высокие ставки отправили экономику в рецессию. Такая жёсткая принципиальная политика ФРС позволила сбить инфляционные ожидания. 

**Какова мораль этой истории?** Политики, начав бороться с безработицей, печатая деньги, немного перепутали причину и следствие. На самом деле никакой чёткой отрицательной зависимости между инфляцией и безработицей в природе не существует. Её наличие в данных является следствием сложного взаимодействия между экономическими агентами. Попытавшись воздействовать на следствие, а не на причину, политики налажали.

В ходе гранд иканам рисёрча мы тоже найдём некоторые закономерности. Из этих закономерностей можно будет сделать либо правильные, либо не очень правильные выводы. Давайте посмотрим на то как это работает.

###  3.1 Пример неправильного вывода: 

> Что что что? Вероятность того, что меня отчислят очень высока? Наличие подписок на мусорные паблики увеличивает эту вероятность? ЭЙ, АВТОР, ПЕРЕДЕЛЫВАЙ ПРОГНОЗЫ! Я ОТПИСАЛСЯ ОТ МДК! 

Нет. Это работает не так. Вы не знаете о том, что по вам делают рисёрч. Вы ведёте в сети какую-то активность. В ходе этой активности вы оставляете некоторый след о своих умственных способностях и особенностях характера. Наличие у вас в подписках МДК — это последствие этой активности. То есть это всего лишь инструмент, который сигнализирует, что у вас __уровень IQ ниже,__ чем у других людей. В ходе рисёрча это становится ясно и модель делает прогноз. 

Как только вы узнали о том, что наличие подписки на МДК выдаёт вас, вы, конечно же решите отписаться от него. Тем не менее, это не увеличит ваши шансы сдать сессию. Просто-напросто вы лишите инструмент, через который модель узнала о вас что-то, своей силы. Не путайте причину и слествие. __Ваш IQ низкий => Вы подписались на мусорный паблик => Модель заметила, что у вас низкий IQ и спрогнозировала высокую вероятность отчисления.__ Но ни в коем случае не наоборот. Отписавшись, вы засорите канал, по которому идёт информация, а не станете умнее. 

В данном примере, уровень интеллекта человека является ненаблюдаемой переменной. Наличие в подписках мусорных пабликов коррелирует с этой скрытой переменной и позволяет нашей модели делать какие-то выводы. Подобные переменные называются **инструментальными.** Инструмент "мусорные подписки" ни в коем случае не является эквивалентом для уровня интеллекта. Тем не менее, он несёт о нём некоторую информацию. В течение исследования мы ни один раз встретимся с понятием инструментальной переменной. 

### 3.2 Пример правильного вывода: 

> Что что что? Вероятность того, что меня отчислят очень высока? Наличие МДК в подписках увеличивает эту вероятность? Плевать я хотел на МДК! Пойду ботать! Ботать это круто! 

Да. Это работает именно так. Вы воздействуете не на следствие того, что у вас низкий IQ, а на причину этого. Именно так и нужно делать. 

### 3.3 Другой пример неправильного вывода: 

> Такс, такс, такс, рисёрч, ага. У меня довольно низкие шансы быть отчисленым. Ну и нафиг всё это ботанье, я пошёл в кино. 

Нет. Это работает не так. То, что модель прогнозирует вам высокий шанс выживания, не означает, что можно ничего не делать. Вы правильно вели себя до этого и усердно трудились. Через разные инструменты, собранные для того, чтобы отследить это (вроде тех же подписок), модель поняла, что на этой сессии вы также будете стараться и приклаыдвать усилия. Работает следующая цепочка. __Он был усерден => он оставил вк след, какой обычно оставляют усердные люди => судя по всему, он будет усердным и в этот раз => низкая вероятность отчисления.__  И ни в коем случае не наоборот! 

### 3.4 Другой пример правильного вывода: 

> Такс, такс, такс, рисёрч, ага. У меня довольно низкие шансы быть отчисленым. Это круто. Можно пойти и как следует поботать. 

Да. Это работает именно так. Здравые мысли здравого человека. 


### 3.5  Ещё один неверный вывод: 

> Ага. Граф друзей. Ага, судя по результатам алгоритма PageRank я не очень авторитетный чувак на своём курсе. Этот алгоритм смотрит на то, кто у меня в друзьях. Добавлю ка я к себе в друзья пару авторитетных пацанов и получу [Признание.](http://bestgames4all.ru/uploads/images/g/t/a/gta_sa_missija_vipolnena.jpg)

Нет, это работает не так. __Вы социально значимый человек  => вы общаетесь с другими социально значимыми людьми => вы, возможно, добавляете их в друзья => у вас высокий PageRank.__  Социальная значимость человека - ненаблюдаемая переменная. Чтобы как-то учесть её при исследовании, приходится искать какие-то другие переменные, которые с ней коррелируют. Как мы уже говорили выше, такие переменные называются инструментами. PageRank, построенный по тому какие именно у человека друзья, коррелирует с его социальной значимостью. 

**Ещё раз, ещё раз.** Он ни в коем случае не является эквивалентом социальной значимости, он лишь несёт о ней некоторую информацию. Искуственно завысив какие-то показатели, по которым можно судить о вашей важности, вы не увеличите свою важность. 

### 3.6  Ещё один неверный вывод: 

> Ага. Граф друзей. Ага, у меня высокий PageRank. Я очень крут. 

Нет, это работает не так. То, что вы попали в топ по одному какому-то алгоритму, не означает, что вы круты.

### 3.7  Ещё один верный вывод: 

> Ага. Граф друзей. Ага, у меня высокий Pagerank. Это интересно. Довольно мило, что данный алгоритм сигнализирует о том, что у меня высокий авторитет. Это не повод зазнаваться. Возможно, у меня просто очень много друзей. Особенно тех, с которыми я не общаюсь и в данной ситуации эта инструментальная переменная оказлась зашумлена. Интересно посмотреть, что скажут другие алгоритмы. 

Да. Это верная логика. Спасибо за здравые мысли. 

### 3.8 Ещё один неверный вывод: 

> Юля очень классная девушка и нравится мне. Вася поставил ей за прошедшие 2 года огромное количество лайков. Пора набить ему морду. Он хочет отбить её у меня. 

Нет, это работает не так. Количество лайков - это интересная метрика. Да, она коррелирует с симпатиями, которые являются ненаблюдаемой переменной, однако бывают люди, которые лайкают всё подряд. Возможно, Юля постит аномально много фоток, а Вася лайкает все фотки подряд. Нужно глубже вникнуть в суть проблемы. 

### 3.9  Ещё один ..... вывод: 

> Юля очень классная девушка и нравится мне. Вася поставил ей за прошедшие 2 года огромное количество лайков. Надо купить Юле цветы и оказывать ей больше знаков внимания. По той теплоте, которую я буду ей дарить, она поймёт как важна для меня. 

Не уверен, но, возможно, да. 

### 3.10 Заключительный ..... вывод: 

> Ваша модель говно. Вы оцеиниваете её на основе всего лишь 300 наблюдений.

Да, данных очень мало. Их всегда мало. Макроэкономисты оценивают свои модели по годовым или квартальным данным за менее чем 20 лет. Тем не менее, это не мешает им получать интересные результаты и лучше понимать то, как именно великий процесс порождения данных задумал окружающий нас мир. Я вовсе не спорю с тем, что 400 наблюдений это мало и о [законе больших чисел](https://ru.wikipedia.org/wiki/Закон_больших_чисел) говорить сложно.

Маленькое количество данных очень сильно связывает нам руки и ограничивает нас классом довольно простых моделей. Мы не можем строить модные случайные леса или нейросетки, однако это вовсе не повод сидеть сложа руки. Три сотни — это на порядок больше, чем 20. Если в данных хоть что-то есть, мы худо-бедно найдём это и используем. Да и к тому же как только кого-нибудь из читателей отчислят, появятся новые данные, и модели можно будет обучить на свеженькой информации.
