## 2 . О Куполе, под которым мы сидим 

Чтобы узнать в красках о том, какой инфомрационный след ты за собой оставляешь, можно посмотреть [видос одного блогера на ютубе.](https://www.youtube.com/watch?time_continue=1&v=swnnpqu7a6s) Рассказано красиво, в эмоцях и немного натянуто.

Откровенно говоря, любому человеку, который работает с даннымми, всё это покажется очевидным. Более того, с частью тезисов даже можно поспорить. Тем не менее, общая картина действительно выглядит примерно так. Машинное обучение, в купе с твоим информационным следом, позволяет любой крупной компании знать о тебе довольно многие вещи.

Например, машинное обучение может помочь [магазину узнать о вашей беременности раньше, чем об этом узнает ваш отец](http://ru.datasides.com/магазин-узнал-о-беременности-девушки/) и даже ненароком сдать вас ему. Оно может помочь вам найти свой собственный [закон пива и подгузников](http://tmstudent.ru/article/keis-pampersy-i-pivo) и повысить прибыль. Более того, оно может помочь вам грамотно сагетировать избирателей и [одержать на выборах победу.](https://theins.ru/politika/38490) Вокруг последних выборов идут довольно жаркие споры. Однако, прочитав книгу [Верховный алгоритм](https://www.ozon.ru/context/detail/id/136780904/), можно узнать, что судьбу США в 2012 году определило именно машинное обучение. Обама оказался первопроходцем в использовании больших данных в политике.

Иногда сбор данных происходит очень неожиданным образом. Например, сегодня социальной сетью вконтакте владеет Mail.Ru Group. Не так давно он [начал эксперементировать с психотипами людей.](https://www.youtube.com/watch?time_continue=1&v=swnnpqu7a6s) В соц сети  было запущено приложение-тест, пройдя который, пользователь мог узнать немного больше о своей психологии. Проходя этот тест, он невольно помогал мэйлу делать разметку выборки. Собрав достаточное количество данных, mail смог обучить модель, которая по соодержимому странчки человека понимает каким психотипом он обладает.

Сегодня данные о психотипах (и другие данные из соц сеток) обученные алгоритмы используют для оценки производительности потенциальных сотрудников. Вполне возможно, что ваше резюме будет отобрано или не отобрано из кипы других именно благодяря такому алгоритму. 

Только что приняли решение почистить свой профиль? [Не думаю, что это поможет.](https://socialdatahub.com/ru/sbor_i_strukturirovanie_dannyih) Представляете себе сколько таких контор, собирающих данные, ведут свою деятельность?  Более того, в последние годы наличка уходит на задний план. Мы оплачиваем практически все транзакции банковскими картами. Это приводит к тому, что банкам доступна детализированная информация о том где и что мы покупаем. Эти данные вы тоже собиратесь как-то почистить?

## 4. Цели и задачи иканам гранд рисёрча 

**Главная цель:** замотивировать тебя ботать. Как луше всего сделать это? Конечно же построить модель, которая будет прогнозировать вероятность отчисления для каждого человека! Идти к этой цели мы будем через решение кучи небольших задач. 

1. Мы соберём данные с контакта. Посмотрим на то какими они бывают и как их можно обрабатывать и использовать. 
2. Мы сделаем предобработку всего этого добра и построим кучу визуализаций. Некоторые из них будут интерактивными.
3. Мы проанализируем кто, кому и сколько ставит лайков. 
4. Мы построим для каждого курса свой граф связей и как следует заанализируем его. 
5. Мы попытаемся найти самых авторитетных людей на иканаме, используя при этом разные концепции. 
6. Мы посмотрим на то какие паблики на иканаме самые популярные, где люди оставляют больше лайков, какие паблики лайкают, но при этом на них не подписываются.
7. Мы выясним что такое тематическое моделирование и как можно автоматически кластеризовать паблики по тематикам. 
8. Мы скачаем кучу фоток и поймём как взаимосвязаны между собой фотошоп и линал.
9. Мы поговорим о нейросетейвой революции и о том как можно обработать картинки и запихнуть к себе в модель. 
10. Мы проверим несколько важных гипотез. Например, довольно давно ходят слухи о том, что когда Козко ведёт матан, с первого курса отчисляют больше людей. Данные помогут нам понять правда это или ложь. 
11. Мы поговорим об асессорах и о мудрости толпы. 
12. Мы посроим несколько моделей, которые спрогнозируют нам вероятность твоего отчисления. Возможно, наша модель даже будет хорошо работать. 

Всё это будет сопровождаться подробным описанием, а также кодом на python с подробными комментариями. Я старался сделать их настолько подробными, чтобы любой человек, который знаком с питоном на начальном уровне, смог понять что в нём происходит и, возможно, адаптировать для своих целей. 

Чёрт возьми! Нас ожидает довольно долгий путь. Однако, прежде чем ступить на скользкую дорожку анализа иканамовских данных, стоит узнать одну поучительную историю. 

## 5. Древняя и поучительная история

В 1958 году была опубликована статья жкономиста Филиппса, в которой он обнаружил достаточно чёткую отрицательную связь между инфляцией и безработицей в Англии за прошедшие 70 лет. Проверка этой работы на американских данных подтвердила наличие взаимосвязи. Так родилась знаменитая кривая Филиппса. Её стали интерпретировать как некую возможность выбора между высокой инфляцией и высокой безработицей. Нужно сократить безработицу? Давайте напечатаем денег и разгоним инфляцию! 

Любому политику безрабтица кажется более значимой социальной проблемой, нежели инфляция, и он хочет её победить любыми доступными средствами. Самым популярным средством по борьбе с безработицей является агрессивная монетарная политика, которая состоит в расширении денежной массы. Именно это и было сделано президентом США Ричардом Никсоном в начале 1970-х гг. в ходе погони за низкой безработицей и высокой инфляцией.

К сожалению, план Никсона удался только наполовину, он добился высокой инфляции, но сбить безработицу не смог. В краткосрочном периоде она действительно упала. Через некоторое время вернулась назад. Кривая Филиппса в этот период времени исчезла из данных. Никой взаимосвязи между безработицей и инфляцией больше не прослеживалось. 

Так или иначе, в течение 70-х все западные страны поэксперементировали с кривой Филиппса. В конечном итоге, 1980-е годы были потрачены на обуздание разогнанной такими методами инфляции. Победа над инфляцией обошлась довольно дорого. Например, Пол Волкер, возглавлявший в этот период времени ФРС США, взвинтил ставку процента до 19%. Конечно же, это отправило экономику в рецессиию, однако инфляционные ожидания были сбиты. 


**Переписать абзац** 

Пацаны, начав использовать кривую Филиппса, как инструмент по борьбе с безработицей, самую малость перепутали причину и следствие. Никакой чёткой отрицательной взаимосвязи между безработицей и инфляцией не было. К её наличию в данных привела именно сложившаяся на тот момент политика.  Именно вся та политика, которая велась до этого, привела к возникновению в данных кривой Филиппса. 

Какова мораль этой истории? В данных была найдена интересная закономерность. Она была неверно истолкована. В итоге это привело к проблемам. В ходе гранд иканм рисёрча мы тоже найдём некоторые закономерности. Из этих закономерностей можно будет сделать либо привильные либо не очень правильные выводы. Давайте посмотрим на то как это работает. 

###  5.1 Пример неправильного вывода: 

> Что что что? Вероятность того, что меня отчислят очень высока? Наличие подписок на мусорные паблики увеличивает эту вероятность? ЭЙ, АВТОР, ПЕРЕДЕЛЫВАЙ ПРОГНОЗЫ! Я ОТПИСАЛСЯ ОТ МДК! 

Нет. Это работает не так. Вы не знаете о том, что по вам делают рисёрч. Вы ведёте в сети какую-то активность. В ходе этой активности вы оставляете некоторый след о своих умственных способностях и особенностях характера. Наличие у вас в подписках МДК это последствие этой активности. Это всего лишь инструмент, который сигнализирует, что у вас уровень IQ __ниже,__ чем у других людей. В ходе рисёрча это становится ясно и модель делает прогноз. 

Как только вы узнали о том, что наличие подписки на МДК выдаёт вас, вы, конечно же решите отписаться от него. Тем не менее, это не увеличит ваши шансы сдать сессию. Просто-напросто вы лишите инструмент, через который модель узнала о вас что-то, своей силы. Не путайте причину и слествие. __Ваш IQ низкий => Вы подписались на мусорный паблик => Модель заметила, что у вас низкий IQ и спрогнозировала высокую вероятность отчисления.__ Но ни в коем случа не наоборот. Отписавшись, вы засорите канал, по которому идёт информация, а не станете умнее. 

### 5.2 Пример правильного вывода: 

> Что что что? Вероятность того, что меня отчислят очень высока? Наличие МДК в подписках увеличивает эту вероятность? Плевать я хотел на МДК! Пойду ботать! Ботать это круто! 

Да. Это работает именно так. Вы воздействуете не на следствие того, что у вас низкий IQ, а на причину этого. Именно так и нужно делать. 

### 5.3 Другой пример неправильного вывода: 

> Такс, такс, такс, рисёрч, ага. У меня довольно низкие шансы быть отчисленым. Ну и нафиг всё это ботанье, я пошёл в кино. 

Нет. Это работает не так. То, что модель прогнозирует вам высокий шанс выживания, не означает, что можно ничего не делать. Вы правильно вели себя до этого и усердно трудились. Через разные инструменты, собранные для того, чтобы отследить это (вроде тех же подписок), модель поняла, что на этой сессии вы также будете стараться и приклаыдвать услия. Работает следующая цепочка. __Он был усерден => он оставил вк след, какой обычно оставляют усердные люди => судя по всему, он будет усердным и в этот раз => низкая вероятность отчисления__  И ни в коем случае не наоборот! 

### 5.4 Другой пример правильного вывода: 

> Такс, такс, такс, рисёрч, ага. У меня довольно низкие шансы быть отчисленым. Это круто. Можно пойти и как следует поботать. 

Да. Это работает именно так. Здравые мысли здравого человека. 


### 5.5  Ещё один неверный вывод: 

> Ага. Граф друзей. Ага, судя по результатам алгоритма PageRank я не очень авторитетный чувак на своём курсе. Этот алгоритм смотрит на то, кто у меня в друзьях. Добавлю ка я к себе в друзья пару авторитетных пацанов и получу [Признание.](http://bestgames4all.ru/uploads/images/g/t/a/gta_sa_missija_vipolnena.jpg)

Нет, это работает не так. __Вы социально значимый человек  => вы общаетесь с другими социально значимыми людьми => вы, возможно, добавляете их в друзья => у вас высокий PageRank.__  Социальная значимость человека - ненаблюдаемая переменная. Чтобы как-то учесть её при исследовании, приходится искать какие-то другие переменные, которые с ней коррелируют. Такие переменные называются инструментами. PageRank, построенный по тому какие именно у человека друзья, коррелирует с его социальной значимостью. 

Ещё раз, ещё раз. Он ни в коем случае не является эквивалентом социальной значимости, он лишь несёт о ней некоторую информацию. Искуственно завысив какие-то показатели, по которым можно судить о вашей важности, вы не увеличите свою важность. 

### 5.6  Ещё один неверный вывод: 

> Ага. Граф друзей. Ага, у меня высокий PageRank. Я очень крут. 

Нет, это работает не так. То, что вы попали в топ по одному какому-то алгоритму, не означает, что вы круты.

### 5.6  Ещё один верный вывод: 

> Ага. Граф друзей. Ага, у меня высокий Pagerank. Это интересно. Довольно мило, что данный алгоритм сигнализирует о том, что у меня высокий авторитет. Это не повод зазнаваться. Возможно, у меня просто очень много друзей. Особенно тех, с которыми я не общаюсь и в данной ситуации эта инструментальная переменная сработала не очень хорошо. Интересно посмотреть, что скажут другие алгоритмы. 

Да. Это верная логика. Спасибо за здравые мысли. 

### 5.7  Ещё один неверный вывод: 

> Юля очень классная девушка и нравится мне. Вася поставил ей за прошедшие 2 года огромное количество лайков. Пора набить ему морду. Он хочет отбить её у меня. 

Нет, это работает не так. Количество лайков это интересная метрика. Да, она коррелирует с симпатиями, которые являются ненаблюдаемой переменной, однако бывают люди, которые лайкают всё подряд. Возможно, Юля постит аномально много фоток, а Вася лайкает все фотки подряд. Нужно глубже вникнуть в суть проблемы. 

### 5.8  Ещё один ..... вывод: 

> Юля очень классная девушка и нравится мне. Вася поставил ей за прошедшие 2 года огромное количество лайков. Надо купить Юле цветы и оказывать ей больше знаков внимания. По той теплоте, которую я буду ей дарить, она поймёт как важна для меня. 

Не уверен, но, возможно, да. 

### 5.9 Заключительный ..... вывод: 

> Ваша модель говно. Вы оцеиниваете её на основе всего лишь 400 наблюдений.

Да, данных очень мало. Их всегда мало. Макроэкономисты оценивают свои модели по годовым или квартальным данным за менее чем 20 лет. Тем не менее, это не мешает им получать интересные результаты и лучше понимать то как именно великий процесс порождения данных задумал окружающий нас мир. Я вовсе не спорю с тем, что 400 наблюдений это мало и о [законе больших чисел](https://ru.wikipedia.org/wiki/Закон_больших_чисел) говорить сложно. Маленькое количество данных вовсе не повод сидеть сложа руки. Четыре сотни — это на порядок больше, чем 20. Да и к тому же как только кого-нибудь из читателей отчислят, появятся новые данные, и модели можно будет обучить на свеженькой информации. 

Тем не менее, я разделяю ваш скепсис. Сомневаться в том, что вам рассказывают и в том, что вы читаете — это правильно. Особенно, когда исследование приходит в точку на подобие вот такой.
