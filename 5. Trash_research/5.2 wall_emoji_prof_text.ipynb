````python
def top_freq_plot(top, title, ikanam_info = ikanam_info, k=30):

    # Сортируем словарик по частотам
    sorted_word_count = sorted(top, key=lambda w: w[1], reverse=True)     
    
    # Делаем срез и по этому срезу строим картиночку 
    example_list = sorted_word_count[:k]
    name = []
    frequency = []
    for i in range(len(example_list)):
        name.append(ikanam_info[example_list[i][0]]['name'])
        frequency.append(example_list[i][1])

    indices = np.arange(len(example_list))
    
    plt.figure(figsize=(22,8))
    plt.bar(indices, frequency)
    plt.xticks(indices, name, rotation='vertical',fontsize=16)
    plt.tight_layout()
    plt.title(title, fontsize = 30)
    pass
````

Как вы могли заметить, многие люди встречаются сразу в нескольких топах. Возможно, у таких людей несколько телефонов. Возможно, такие люди берут погонять мобилу у друзей, подруг или просто-напросто работают гопниками. Также возможно, что они просто-напросто в какой-то момент своей жизни поменяли марку телефона. Давайте попробуем найти людей, которые пересели с андроида на айфон и наоборот. 

Как понять, что такое произошло? Человек долгое время постил с одного телефона, потом резко начал постить с другого и не смог остановиться. Ясное дело, что для идентификации подобных событий нам понадобится длиная история стен. Будем руководствоваться следующей эвристикой (интуитивны правилом, которое взято с потолка):

> Если человек сделал 10 постов с одной марки телефона, а после сделал 10 постов с другой марки телефона, значит он сменил телефон.

Звучит логично. Попробуем эту эвристику на практике. Все пропуски в платформах (это означает, что пост оставлен с компа), а также ipad и тп мы выбросим из рассмотрения. Считать будем только андроид, айфон и виндус-фон. 

````python
phone_df = df_wall[['uid','platform','date']].dropna()
int_uids = list(set(phone_df['uid'].get_values()))  # Интересующие нас id 
````

````python
def post_platform(uid, sort = True, gamma = 10):
    # Отобрали текущего юзера
    cur_user_df = phone_df[phone_df.uid == uid]

    platforms = cur_user_df.platform.get_values()  # вытащили вектор из айфонов, андроидов и тп 
    dates = cur_user_df.date.get_values()          # отдельный вектор с датами

    base_platform = platforms[0]                   # с чего был оставлен первый пост 

    posts_posled = [ ]  # Сюда сохраняем имена устройств, с которых делались посты. 
    ph_cnt = 1          # Счётчик постов 
    for post,date in zip(platforms[1: ],dates[1:]):
        if post == base_platform:  # увеличиваем счётчик, если посты с одного устройства 
            ph_cnt += 1
        else: 
            # если последовательность надломлена, то сбрасываем счётчик и начинаем собирать её заново
            posts_posled.append((base_platform, ph_cnt, date))
            ph_cnt = 1 
            base_platform = post

    # Отсекаем все последовательности, которые ниже порогового числа записей: 
    # Также отсекаем инстаграм и тп 
    posts_posled = [item for item in posts_posled if (item[1] > gamma)&(item[0] in ['iphone', 'android', 'wphone'])]
    
    if sort:
        return sorted(posts_posled, key=lambda w: w[1], reverse=True)
    else:
        return posts_posled
````

````python
post_platform(int_uids[0])
````

````python
potential_changers = [ ]
for uid in int_uids: 
    y = post_platform(uid)
    if len(y) > 1:
        if(len(set([item[0] for item in y])) > 1):
            potential_changers.append((ikanam_info[uid]['name'], y))
````

````python
len(potential_changers)
````

Отлично! С помощью нашей эвристики мы нашли 24 человека, которые, скорее всего, меняли телефоны. Давайте посмотрим на этих людей.  В векторе, который мы будем выводить, будет написана:


````python
for item in potential_changers:
    print(item[0],'менял(а) телефон:')
    for jtem in item[1]:
        print(jtem)
    print('\n')
````

# 2. Кластеризация эмодзи

Попробуем ради угара замутить кластеризацию людей по эмодзи на их стене. Эмодзи-анализ ёпт. Вот такие у нас есть эмодзи-следы. 

````python
em = [em for em in df_wl.wall_emoji_trace.dropna().get_values() if len(em) > 0]
em[0]
````

Раздробим эмодзи-следы на отдельные эмодзи.

````python
from nltk.tokenize import RegexpTokenizer 
tokenizer = RegexpTokenizer('.')
X_emoji = [tokenizer.tokenize(mm) for mm in em]
````

Запилим словарь эмодзи и корпус эмодзи. Всё как при кластеризации групп.

````python
from gensim import corpora, models
dictionary = corpora.Dictionary(X_emoji)                 # составляем словарь из терминов
corpus = [dictionary.doc2bow(text) for text in X_emoji]  # составляем корпус документов
````

Каждому эмодзи соответствует свой номер.

````python
dictionary.token2id['😜']
````

Всего в выборке 2303 уникальных эмодзи. В эмодзи затесались иероглифы. Наверное это не особо удивительно. 

````python
len(dictionary)
````

Можно отфильтровать словарик от стоп-эмодзи и слишком редких эмодзи, но это приведёт к тому, что всё выродится в один кластер. Код для фильтрации я здесь конечно оставлю, но закоменчу. 

````python
ing_2_50 = [token_id for token_id,freq in dictionary.dfs.items() if (freq < 10)|(freq > 180)]

dictionary.filter_tokens(ing_2_50)
print(len(dictionary))
corpus = [dictionary.doc2bow(text) for text in X_emoji]
````

Выделяем пять эмодзи-тем. 

````python
%%time
np.random.seed(42)
ldamodel = models.ldamodel.LdaModel(corpus, id2word=dictionary,num_topics=5, passes=20)
````

Смотрим на термины по каждой теме. Видим какой-то бред. 

````python
topics = ldamodel.show_topics(num_topics=10, num_words=20,formatted=False)
i = 0
for top in topics:
    print(i, [item[0] for item in top[1]],'\n')
    i += 1
````

# 3. Кластеризация информации из профилей 

Подгрузили инфу из профилей. 

````python
df_prof = pd.read_csv('profile_about_v18-12-17.csv', sep = ',', index_col=0)
print(df_prof.shape)
df_prof.head()
````

Заполнили пропуски и объединили все строки в единый текст. Посмотрим на первую и предпоследнюю записи в профилях. Почему именно на них? Они доставляют. 

````python
df_prof = df_prof.fillna('')
prof_info = [''.join(df_prof.iloc[i].get_values()[:-1]).lower() for i in range(df_prof.shape[0])]
prof_full = [item for item in prof_info if len(item) > 0]
print(len(prof_full))
print(prof_full[0])
print(prof_full[-2])
````

Мужские шишки. С-с-с-c-c. Мурлыкала. Хы-хы-хы. 

````python
tokenizer = RegexpTokenizer(r'\w+')
X_prof = [tokenizer.tokenize(txt) for txt in prof_full]

dictionary = corpora.Dictionary(X_prof)                 # составляем словарь из терминов
len(dictionary)
````

Словарь вышел слишком большим. Уменьшаем его. 

````python
ing_2_50 = [token_id for token_id,freq in dictionary.dfs.items() if (freq < 2)|(freq > 5)]
len(ing_2_50)

dictionary.filter_tokens(ing_2_50)
print(len(dictionary))
corpus = [dictionary.doc2bow(text) for text in X_prof]
````

Учим модель с 5 темами. 

````python
%%time
np.random.seed(42)
ldamodel = models.ldamodel.LdaModel(corpus, id2word=dictionary, num_topics=5, passes=20)
````

Смотрим чё получилось. 

````python
topics = ldamodel.show_topics(num_topics=10, num_words=20,formatted=False)
i = 0
for top in topics:
    print(i, [item[0] for item in top[1]],'\n')
    i += 1
````

Конечно же чуть менее, чем ничего.  Вроде бы во вторую темы выделились околоанглийские слова. Давайте чтоль сгенерим описание для нового профиля. 

````python
def generate_content(model, num_words):
    theta = np.random.dirichlet(model.alpha)
    for i in range(num_words):
        t = np.random.choice(np.arange(model.num_topics), p=theta)
        topic = model.show_topic(0, topn=model.num_terms)
        topic_distr = [x[1] for x in topic]
        terms = [x[0] for x in topic]
        # Тут есть костыль. Он делается, чтобы вероятности в сумме давали 1, а не 0.9999999
        w = np.random.choice(terms, p=1/sum(topic_distr)*np.array(topic_distr))
        print(w)

generate_content(ldamodel, 20)
````

В принципе получилось что-то близкое к реальности. Кто в 2к18 заполняет профили? Хватит трэша. Пришло время закругляться.
